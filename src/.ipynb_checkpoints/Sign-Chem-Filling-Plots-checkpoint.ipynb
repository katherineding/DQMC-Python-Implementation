```python
plotroot="hallplots/"
```

```python
Nx = "6"; Ny = "6"; nt = "0.85";
nflux_list = ['0','1','4','8','12','16'];
nF = len(nflux_list)
```

```python
for k in range(nF):
    nflux = nflux_list[k]
    pathroot = f"/home/katherine/Documents/GradWork/Devereaux/Data-DQMC/{Nx}x{Ny}/nflux{nflux}/n{nt}/"
    plt.figure();
    plt.title(f'{Nx}x{Ny} nflux={nflux} n = {nt}')
    betas_list, Us_list, mu_dict = da.readmu(pathroot,"best_mu.txt",show=True)
    nT = len(betas_list); nU = len(Us_list)
    beta_list = np.array(list(map(float,betas_list)))
    U_list = np.array(list(map(float,Us_list)))
    border = np.argsort(beta_list); Uorder = np.argsort(U_list)
    #plt.savefig(f'{plotroot}{Nx}x{Ny}_n{nt}_nflux{nflux}_chemical_potential.png',dpi=300,bbox_inches="tight")
    #plt.close()
```

```python
s_arr = np.empty((nT,nU,nF),dtype=complex) * np.nan
se_arr = np.empty((nT,nU,nF),dtype=float) * np.nan
d_arr = np.empty((nT,nU,nF),dtype=complex) * np.nan
de_arr = np.empty((nT,nU,nF),dtype=float) * np.nan

for k in range(nF):
    nflux = nflux_list[k]
    pathroot = f"/home/katherine/Documents/GradWork/Devereaux/Data-DQMC/{Nx}x{Ny}/nflux{nflux}/n{nt}/"
    betas_list, Us_list, mu_dict = da.readmu(pathroot,"best_mu.txt")
    beta_list = np.array(list(map(float,betas_list)))
    U_list = np.array(list(map(float,Us_list)))
    #reorder beta and U lists
    border = np.argsort(beta_list); Uorder = np.argsort(U_list)
    for i in border:
        betas = betas_list[i]
        for j in Uorder:
            Us = Us_list[j]
            mu = mu_dict[(betas,Us)]
            try:
                path= pathroot + f"beta{betas}_U{Us}_mu{mu:.3f}/"
                s_arr[i,j,k], se_arr[i,j,k] = da.full_sign(path,uneqlt=True)
                d_arr[i,j,k], de_arr[i,j,k] = da.full_density(path)
            except IndexError as e:
                print(e)
```

```python
colors = plt.cm.plasma(np.linspace(1,0,nT))
for k in range(nF):
    nflux=nflux_list[k]
    plt.figure()
    plt.title(f"{Nx}x{Ny} nflux={nflux} n={nt} sign real part")
    n=0
    for i in border:
        beta = beta_list[i]
        plt.errorbar(U_list[Uorder],s_arr.real[i,:,k][Uorder],yerr = se_arr[i,:,k][Uorder],\
                     fmt='.:',label=f"beta={beta:.3g}",color=colors[n])
        n+=1
    plt.legend(loc="best")
    plt.xlabel("U/t")
    plt.ylim(2e-3,1.2)
    plt.yscale("log")
```

```python
colors = plt.cm.plasma(np.linspace(1,0,nU))
for j in Uorder:
    Us = Us_list[j]
    plt.figure()
    plt.title(f"{Nx}x{Ny} U={Us} n={nt} sign real part")
    n=0
    for i in border:
        beta = beta_list[i]
        plt.errorbar(list(map(float,nflux_list)),s_arr.real[i,j,:], yerr = se_arr[i,j,:],\
                     fmt='.:',label=f"beta={beta:.3g}",color=colors[n])
        n+=1
    plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
    plt.xlabel("nflux")
    plt.ylim(2e-3,1.2)
    plt.yscale("log")
```

```python
colors = plt.cm.plasma(np.linspace(1,0,nT))
for k in range(nF):
    nflux=nflux_list[k]
    plt.figure()
    plt.title(f"{Nx}x{Ny} nflux={nflux} n={nt} density real part")
    n=0
    for i in border:
        beta = beta_list[i]
        plt.errorbar(U_list[Uorder],d_arr.real[i,:,k][Uorder],yerr = de_arr[i,:,k][Uorder],\
                     fmt='.:',label=f"beta={beta:.3g}",color=colors[n])
        n+=1
    plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
    plt.xlabel("U/t")
    plt.ylim(float(nt)-5e-3,float(nt)+5e-3)
```
