```python
for k in range(nF):
    nflux = nflux_list[k]
    if tp =="0":
        pathroot = f"/home/{user}/Documents/GradWork/Devereaux/Data-DQMC/{Nx}x{Ny}/nflux{nflux}/n{nt}/"
    else:
        pathroot = f"/home/{user}/Documents/GradWork/Devereaux/Data-DQMC/{Nx}x{Ny}_tp{tp}/nflux{nflux}/n{nt}/"
    plt.figure();
    plt.title(tt_prefix + f" nflux={nflux_list[k]}")
    betas_list, Us_list, mu_dict = da.readmu(pathroot,"best_mu.txt",show=True)
    plt.show()
    

if nt == "0.9" or nt =="0.85": #only relevant for 6x6 quirks
    muroot = f"/home/{user}/Documents/GradWork/Devereaux/Data-DQMC/{Nx}x{Ny}/nflux2/n{nt}/"  
else:
    muroot = f"/home/{user}/Documents/GradWork/Devereaux/Data-DQMC/{Nx}x{Ny}/nflux1/n{nt}/"     
betas_list, Us_list, mu_dict = da.readmu(muroot,"best_mu.txt",show=False)
nT = len(betas_list); nU = len(Us_list)
beta_list = np.array(list(map(float,betas_list)))
U_list = np.array(list(map(float,Us_list)))
border = np.argsort(beta_list); Uorder = np.argsort(U_list)
```

```python
print("nflux_list:",nflux_list)
print("betas_list:",betas_list)
print("Us_list:",Us_list)
```

# Equal time measurements of size Nx x Ny

```python
meas_list = ["zz",'xx','nn','pair_sw','g00']
```

```python
### Method 1 : Dictonary. This reduces file reads, 
###    but only saves progress after all have been completed
###

#initialize dictionaries
data_dict = {}; 
dataerr_dict = {}; 
for meas_name in meas_list:
    data_dict[meas_name] = np.full((nT,nU,nF,int(Nx),int(Ny)),np.nan,dtype=complex)
    dataerr_dict[meas_name] = np.full((nT,nU,nF,int(Nx),int(Ny)),np.nan,dtype=float)


for k in range(nF):
    nflux = nflux_list[k]
    muroot = f"/home/{user}/Documents/GradWork/Devereaux/Data-DQMC"+\
        f"/{Nx}x{Ny}/nflux{nflux}/n{nt}/"
    _, _, mu_dict = da.readmu(muroot,"best_mu.txt",show=False)
    for j in Uorder:
        Us = Us_list[j]
        for i in border:
            betas=betas_list[i]
            try: 
                mu = mu_dict[(betas,Us)]
                dpath = muroot + f"beta{betas}_U{Us}_mu{mu:.3f}/"
                dm, de = da.eqlt_meas_Nx_Ny(dpath,meas_list)
                print(dpath)
                for meas_name in meas_list:
                    data_dict[meas_name][i,j,k] = dm[meas_name];
                    dataerr_dict[meas_name][i,j,k] = de[meas_name];
            except:
                pass
            


            
```

```python
for meas_name in meas_list:
    np.save(resultroot+f"{meas_name}/Nx{Nx}_Ny{Ny}_tp{tp}_n{nt}_mu_dict.npy",mu_dict)
    np.save(resultroot+f"{meas_name}/Nx{Nx}_Ny{Ny}_tp{tp}_n{nt}_{meas_name}.npy",data_dict[meas_name])
    np.save(resultroot+f"{meas_name}/Nx{Nx}_Ny{Ny}_tp{tp}_n{nt}_{meas_name}err.npy",dataerr_dict[meas_name])
    np.save(resultroot+f"{meas_name}/Nx{Nx}_Ny{Ny}_tp{tp}_n{nt}_nflux.npy",nflux_list)
    np.save(resultroot+f"{meas_name}/Nx{Nx}_Ny{Ny}_tp{tp}_n{nt}_Us.npy",Us_list)
    np.save(resultroot+f"{meas_name}/Nx{Nx}_Ny{Ny}_tp{tp}_n{nt}_betas.npy",betas_list)
    data_ld = np.load(resultroot+f"{meas_name}/Nx{Nx}_Ny{Ny}_tp{tp}_n{nt}_{meas_name}.npy")
    num_ok = np.count_nonzero(~np.isnan(data_ld))
    print(f"{int(num_ok)}/{data_ld.size} = {int(num_ok)/data_ld.size:.3g} of data is saved in directory {resultroot}{meas_name}/")
    print(f"{meas_name}_ld.shape = {data_ld.shape}")
```
