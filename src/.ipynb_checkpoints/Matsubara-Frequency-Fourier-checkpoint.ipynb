```python
def matsubara_freq(path,bs = 1):
    #filling and sign info
    if da.info(path,uneqlt=False,show=True) > 0: 
        return np.nan
    beta,Nx,Ny,bps,N, L, U,dt = \
        util.load_firstfile(path,"metadata/beta","metadata/Nx","metadata/Ny","metadata/bps",\
                            "params/N","params/L","metadata/U","params/dt")
    ns, s, jj = util.load(path,
        "meas_uneqlt/n_sample", "meas_uneqlt/sign", "meas_uneqlt/jj")
    #reshape jj into more intuitive shape
    jj = np.reshape(jj,(-1,Nx,Ny,bps,bps,L),order='F')
    jj = np.transpose(jj,(0,5,4,3,1,2))
    

    #some bins are complete, work with completed ones
    mask = ns == ns.max(); nbin = mask.sum()
    ns, s, jj= ns[mask],s[mask], jj[mask]
    
    jjq0 = jj.sum((-1,-2))   #take q == 0
    #forumlas specific to tp = 0, bps=2 case
    jxjxq0 = jjq0[:,:, 0, 0]; jyjyq0 = jjq0[:,:, 1, 1]
    jxjyq0 = jjq0[:,:, 1, 0]; jyjxq0 = jjq0[:,:, 0, 1]
    #look at profiles
    w = np.arange(L)*2*np.pi/beta
    
        
    #chi = <(jx + i*jy) (jx-i*jy)> 
    chiq0_asymm = 0.5*(jxjyq0 - jyjxq0)/np.mean(s)
    chiq0_symm =  -0.5*(jxjxq0 + jyjyq0)/np.mean(s)
    
    asym_col0 = np.reshape(chiq0_asymm[:,0],(nbin,1))
    sym_col0 = np.reshape(chiq0_symm[:,0],(nbin,1))
    ca = np.concatenate((chiq0_asymm,-asym_col0),axis=1)
                
    sm = np.mean(chiq0_symm,axis=0)
    am = np.mean(chiq0_asymm,axis=0)
    cm = np.mean(ca,axis=0)
    mean_midpt = (cm[:-1] + cm[1:])*0.5
    
    plt.figure()
    plt.plot(w,fft_midpt(beta,L,mean_midpt).real,'.:',label="real")
    plt.plot(w,fft_midpt(beta,L,mean_midpt).imag,'+',label='imag')
    plt.title(tt_prefix + fr" $\beta$ = {beta:.3g} U = {U}")
    plt.xlabel(r"$\omega$")
    plt.ylabel("DFT: midpoint method")
    plt.axhline(y = 0,ls="-.",color="k")
    plt.axvline(x = L*np.pi/beta,ls="--",color='r',label="Nyquist frequency")
    plt.legend(loc="best")
    plt.show()
    
    plt.figure()
    plt.errorbar(np.arange(L+1)*dt,cm.real, yerr = np.std(ca.real,axis=0),fmt='s:',label="data real")
    plt.errorbar(np.arange(L+1)*dt,cm.imag, yerr = np.std(ca.imag,axis=0),fmt='^:',label="data imag")
    plt.plot((np.arange(L)+0.5)*dt,ifft_midpt(beta,L,fft_midpt(beta,L,mean_midpt)).real,'.',label="recon real")
    plt.plot((np.arange(L)+0.5)*dt,ifft_midpt(beta,L,fft_midpt(beta,L,mean_midpt)).imag,'.',label="recon imag")
    plt.title(tt_prefix + fr" $\beta$ = {beta:.3g} U = {U}")
    plt.xlabel(r"$\tau$")
    plt.ylabel("midpoint method reconstruction")
    plt.axhline(y = 0,ls="-.",color="k")
    plt.axvline(x = beta/2,ls="--",color='r',label=r"$\tau = \beta/2$")
    plt.legend(loc="best")
    plt.show()
   
    y = fft_midpt(beta,L,am)
    
    
    return y[1].imag/w[1]
```

```python
def fft_midpt(beta,L,Gmidpt):
    out = np.full(L,np.nan,dtype=complex)
    for n in range(L): 
        out[n] = beta/L*np.exp(1j*n*np.pi/L)*np.sum(np.exp(2j*np.pi/L * n * np.arange(L)) * Gmidpt)
    return out
```

```python
def ifft_midpt(beta,L,Gleft):
    out = np.full(L,np.nan,dtype=complex)
    for m in range(L): 
        out[m] = 1/beta*np.sum(np.exp(-2j*np.pi* m /L * np.arange(L)) * np.exp(-1j*np.pi/L * np.arange(L)) * Gleft)
    return out
```

```python
from scipy.interpolate import UnivariateSpline, InterpolatedUnivariateSpline
from scipy.integrate import quad
```

```python
def spline_test(path,bs = 1):
    #filling and sign info
    if da.info(path,uneqlt=False,show=False) > 0: 
        return np.nan
    beta,Nx,Ny,bps,N, L, U,dt = \
        util.load_firstfile(path,"metadata/beta","metadata/Nx","metadata/Ny","metadata/bps",\
                            "params/N","params/L","metadata/U","params/dt")
    ns, s, jj = util.load(path,
        "meas_uneqlt/n_sample", "meas_uneqlt/sign", "meas_uneqlt/jj")
    #reshape jj into more intuitive shape
    jj = np.reshape(jj,(-1,Nx,Ny,bps,bps,L),order='F')
    jj = np.transpose(jj,(0,5,4,3,1,2))
    

    #some bins are complete, work with completed ones
    mask = ns == ns.max(); nbin = mask.sum()
    ns, s, jj= ns[mask],s[mask], jj[mask]
    
    jjq0 = jj.sum((-1,-2))   #take q == 0
    #forumlas specific to tp = 0, bps=2 case
    jxjxq0 = jjq0[:,:, 0, 0]; jyjyq0 = jjq0[:,:, 1, 1]
    jxjyq0 = jjq0[:,:, 1, 0]; jyjxq0 = jjq0[:,:, 0, 1]
    #look at profiles
    w = np.arange(L)*2*np.pi/beta
    
        
    #chi = <(jx + i*jy) (jx-i*jy)> 
    chiq0_asymm = 0.5*(jxjyq0 - jyjxq0)/np.mean(s)
    chiq0_symm =  -0.5*(jxjxq0 + jyjyq0)/np.mean(s)
    
    asym_col0 = np.reshape(chiq0_asymm[:,0],(nbin,1))
    sym_col0 = np.reshape(chiq0_symm[:,0],(nbin,1))
    ca = np.concatenate((chiq0_asymm,-asym_col0),axis=1)
                
    sm = np.mean(chiq0_symm,axis=0)
    am = np.mean(chiq0_asymm,axis=0)
    cm = np.mean(ca,axis=0)
    mean_midpt = (cm[:-1] + cm[1:])*0.5
    ttest = np.linspace(0,beta,101)
    
    spl = InterpolatedUnivariateSpline(np.arange(L+1)*dt,cm.imag,ext=2,check_finite=True)
    
    y = np.full(L,np.nan,dtype=float)
    yerr = np.full(L,np.nan,dtype=float)
    z = np.full(L,np.nan,dtype=float)
    zerr = np.full(L,np.nan,dtype=float)
    for n in range(L):
        f = lambda tau : -np.sin(2*n*np.pi/beta*tau)*spl(tau)
        g = lambda tau : np.cos(2*n*np.pi/beta*tau)*spl(tau)
        y[n],yerr[n] = quad(f,0,beta)
        z[n],zerr[n] = quad(g,0,beta)
    '''    
    plt.figure()
    plt.title(tt_prefix + fr" $\beta$ = {beta:.3g} U = {U}")
    plt.errorbar(np.arange(L+1)*dt,cm.imag,yerr=np.std(ca.imag,axis=0),fmt='s',label="data")
    plt.plot(ttest,spl(ttest),label="spline fit")
    plt.axhline(y=0,ls="-.",color='k')
    plt.legend(loc="best")
    plt.show()
    
    plt.figure()
    plt.errorbar(w,y,yerr = yerr,fmt = '.:',label="real")
    plt.errorbar(w,z,yerr = zerr,fmt = '+',label='imag')
    plt.title(tt_prefix + fr" $\beta$ = {beta:.3g} U = {U}")
    plt.xlabel(r"$\omega$")
    plt.ylabel(r"$\chi_{xy}(i\omega_n)$")
    plt.axhline(y = 0,ls="-.",color="k")
    plt.axvline(x = L*np.pi/beta,ls="--",color='r',label="Nyquist frequency")
    plt.legend(loc="best")
    #plt.savefig(f'{plotroot}{Nx}x{Ny}_tp{tp}_n{nt}_U{U}_beta{beta:.3g}_FT.png',dpi=300,bbox_inches="tight")
    plt.show()'''
   
    
    return y[1]/w[1]
```
