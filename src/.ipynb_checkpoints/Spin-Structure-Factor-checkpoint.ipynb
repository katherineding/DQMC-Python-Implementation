```python
import numpy as np
import matplotlib.pyplot as plt
import os
user = os.environ.get('USER')
import data_analysis as da
import spin_struct
%load_ext autoreload
```

```python
%autoreload 2
```

```python
plotroot="hallplots/spin-struct-fact/"
resultroot="halldata/spin-struct-fact/"
```

```python
Nx = "6"; Ny = "6"; nt = "0.95"; tp='0'; N = int(Nx) * int(Ny)
nflux_list = ['0','1','2','3','4','6','8']
nf_arr = np.array(list(map(float,nflux_list)))
nF = len(nflux_list)
tt_prefix = f"{Nx}x{Ny} t'={tp} n={nt}"
```

```python
for k in range(nF):
    nflux = nflux_list[k]
    pathroot = f"/home/{user}/Documents/GradWork/Devereaux/Data-DQMC/{Nx}x{Ny}/nflux{nflux}/n{nt}/"
    plt.figure();
    plt.title(tt_prefix + f" nflux={nflux_list[k]}")
    betas_list, Us_list, mu_dict = da.readmu(pathroot,"best_mu.txt",show=True)
    plt.show()
    
if nt == "0.95" or nt == "1.0":
    muroot = f"/home/{user}/Documents/GradWork/Devereaux/Data-DQMC/{Nx}x{Ny}/nflux1/n{nt}/"      
else:
    muroot = f"/home/{user}/Documents/GradWork/Devereaux/Data-DQMC/{Nx}x{Ny}/nflux2/n{nt}/"  
betas_list, Us_list, mu_dict = da.readmu(muroot,"best_mu.txt",show=False)
nT = len(betas_list); nU = len(Us_list)
beta_list = np.array(list(map(float,betas_list)))
U_list = np.array(list(map(float,Us_list)))
border = np.argsort(beta_list); Uorder = np.argsort(U_list)
```

```python
nw = 200
print("nflux_list:",nflux_list)
print("betas_list:",betas_list)
print("Us_list:",Us_list)
bs = 1 #TODO: bootstrap error estimation
```

```python
dim_x = int(Nx)//2+1;dim_y = int(Ny)//2 + 1
S_collect = np.full((nT,nU,nF,bs,dim_x,dim_y,nw),np.nan)
S_omega_collect = np.full((nT,nU,nF,nw),np.nan)
S_domega_collect = np.full((nT,nU,nF,nw),np.nan)

for k in range(nF):
    nflux = nflux_list[k]
    pathroot = f"/home/{user}/Documents/GradWork/Devereaux/Data-DQMC/{Nx}x{Ny}/nflux{nflux}/n{nt}/"
    _, _, mu_dict = da.readmu(pathroot,"best_mu.txt")
    for j in Uorder[1:]:
        Us = Us_list[j]
        for i in border[6:]:
            betas = betas_list[i]
            try:
                mu = mu_dict[(betas,Us)]
            except KeyError as e:
                print(e);
                continue
            path = pathroot + f"beta{betas}_U{Us}_mu{mu:.3f}/"
            _, chi, w, dw = spin_struct.struct_fact(path,nw=nw,bs=bs,method = "BT",anneal_arr = None,checks=False)
            S_collect[i,j,k] = chi
            S_omega_collect[i,j,k] = w
            S_domega_collect[i,j,k] = dw
```

```python
np.save(resultroot+f"Nx{Nx}_Ny{Ny}_n{nt}_tp{tp}_S.npy",S_collect)
np.save(resultroot+f"Nx{Nx}_Ny{Ny}_n{nt}_tp{tp}_Somega.npy",S_omega_collect)
np.save(resultroot+f"Nx{Nx}_Ny{Ny}_n{nt}_tp{tp}_Sdomega.npy",S_domega_collect)
```
