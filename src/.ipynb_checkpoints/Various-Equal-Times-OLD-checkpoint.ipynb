```python
def get_mz(path):
    '''Load measurements from SHERLOCK DQMC runs and look at them'''
    Nx, Ny, bps,U, tp, nflux, mu, beta, N, L, dt = \
    util.load_firstfile(path,'metadata/Nx','metadata/Ny','metadata/bps','metadata/U',\
                        'metadata/t\'','metadata/nflux','metadata/mu','metadata/beta',\
                        'params/N','params/L','params/dt')
    
    n_sample_eqlt, sign_eqlt, density, g00, double_occ  = \
    util.load(path,"meas_eqlt/n_sample", "meas_eqlt/sign",'meas_eqlt/density',\
              "meas_eqlt/g00", "meas_eqlt/double_occ")
        
    
    f=lambda s, o1, o2: ((o1.T-2*o2.T)/s.T).T
    #jackknife is fidgety about matrix shapes
    mz = util.jackknife(sign_eqlt,np.squeeze(density),np.squeeze(double_occ),f = f)
    #print(mz)
    
    return mz
```

```python
def get_energy(path):
    '''Load measurements from SHERLOCK DQMC runs and look at them'''
    Nx, Ny, bps,U, tp, nflux, mu, beta, N, L, dt ,degen_ij= \
    util.load_firstfile(path,'metadata/Nx','metadata/Ny','metadata/bps','metadata/U',\
                        'metadata/t\'','metadata/nflux','metadata/mu','metadata/beta',\
                        'params/N','params/L','params/dt','params/degen_ij')
    #print(f'Interactions:U = {U}, t = 1, tp = {tp}, mu = {mu}, bps={bps}, nflux = {nflux}')
    #print(f'Temerature: beta = {beta} dt = {dt} L = {L}')
    
    n_sample, sign, density, g00, double_occ  = \
    util.load(path,"meas_eqlt/n_sample", "meas_eqlt/sign",'meas_eqlt/density',\
              "meas_eqlt/g00", "meas_eqlt/double_occ")
    g00 = np.reshape(g00,(-1,Nx,Ny),order='F')
#     plt.figure()
#     plt.imshow(np.mean(g00,axis=0))
#     plt.colorbar()
#     plt.show()
    #checked against https://arxiv.org/abs/cond-mat/9612132v2 FIG.2
    #energy zero point differ for U != 0, but variation looks right
    #t1 = (g00[:,0,1]+g00[:,1,0]+g00[:,0,Ny-1]+g00[:,Nx-1,0]) 
    t1 = 2*(g00[:,0,1]+g00[:,1,0]+g00[:,0,Ny-1]+g00[:,Nx-1,0]) 
#     t1[:] = 0
    t3 = U*(double_occ-0.5*density)
    
    f= lambda s, o1, o3: ((o1.T+o3.T)/s.T).T
    energy_jk = util.jackknife(sign,t1,np.squeeze(t3),f=f)
    print(energy_jk)
    
    return energy_jk[0]+1/4*U,energy_jk[1]
```

```python
def get_struct_factors(path):
    '''Load measurements from SHERLOCK DQMC runs and look at them'''
    Nx, Ny, bps,U, tp, nflux, mu, beta, N, L, dt ,degen_ij= \
    util.load_firstfile(path,'metadata/Nx','metadata/Ny','metadata/bps','metadata/U',\
                        'metadata/t\'','metadata/nflux','metadata/mu','metadata/beta',\
                        'params/N','params/L','params/dt','params/degen_ij')
    #print(f'Interactions:U = {U}, t = 1, tp = {tp}, mu = {mu}, bps={bps}, nflux = {nflux}')
    #print(f'Temerature: beta = {beta} dt = {dt} L = {L}')
    
    n_sample, sign , zz,nn = util.load(path,"meas_eqlt/n_sample", "meas_eqlt/sign",'meas_eqlt/zz','meas_eqlt/nn')
    
    zz = np.reshape(zz,(-1,Nx,Ny),order='F')
    nn = np.reshape(nn,(-1,Nx,Ny),order='F')
    
    kpi = np.array([np.pi,np.pi]) #wave vector
    phasemat = np.zeros((Nx,Ny),dtype=float)
    for loc in product(range(Nx),range(Ny)):
        phase = np.dot(kpi,np.array(loc))
        phasemat[loc[0],loc[1]] = phase
        

    #print(np.exp(1j*phasemat).real)
    cdw = np.sum(np.exp(1j*phasemat).real*nn,axis = (1,2))
    sdw = np.sum(np.exp(1j*phasemat).real*zz,axis = (1,2))
    
    cdw_jk = util.jackknife(sign,cdw)
    sdw_jk = util.jackknife(sign,sdw)
    #print(sdw_jk.shape)
    
    return cdw_jk, sdw_jk
```

```python
def get_S_k(path,meas_name):
    '''Load measurements from SHERLOCK DQMC runs and look at them'''
    Nx, Ny, bps,U, tp, nflux, mu, beta, N, L, dt ,degen_ij= \
    util.load_firstfile(path,'metadata/Nx','metadata/Ny','metadata/bps','metadata/U',\
                        'metadata/t\'','metadata/nflux','metadata/mu','metadata/beta',\
                        'params/N','params/L','params/dt','params/degen_ij')
    #print(f'Interactions:U = {U}, t = 1, tp = {tp}, mu = {mu}, bps={bps}, nflux = {nflux}')
    #print(f'Temerature: beta = {beta} dt = {dt} L = {L}')
    dm,de = da.full_density(path)
    tt = fr"{Nx}x{Ny} nflux={nflux} n={dm.real:.3g} t'={tp} U={U} $\beta$={beta:.3g}"
    n_sample, sign , dat = \
    util.load(path,"meas_eqlt/n_sample", "meas_eqlt/sign",f'meas_eqlt/{meas_name}')
    
    dat = np.reshape(dat,(-1,Nx,Ny),order='F')
    plt.figure()
    plt.imshow(dat.mean(0)/np.mean(sign));plt.colorbar()
    plt.title(tt+f"real space {meas_name}")
    plt.xlabel("y")
    plt.ylabel("x")
    print(np.sum(dat.mean(0)/np.mean(sign)))
    
    #all allowed values of momentum, [-pi,pi]
    kx_arr = np.linspace(-np.pi,np.pi,Nx+1)
    ky_arr = np.linspace(-np.pi,np.pi,Ny+1)
    sus_val = np.zeros((len(kx_arr),len(ky_arr)))
    sus_err = np.zeros((len(kx_arr),len(ky_arr)))
    for kid in product(range(kx_arr.shape[0]),range(ky_arr.shape[0])):
        kx = kx_arr[kid[0]]
        ky = ky_arr[kid[1]]
        k = np.array((kx,ky))#wave vector
        phasemat = np.zeros((Nx,Ny),dtype=float)
        for loc in product(range(Nx),range(Ny)):
            phase = np.dot(k,np.array(loc))
            phasemat[loc[0],loc[1]] = phase
        

        #print(np.exp(1j*phasemat).real)
        sus = np.sum(np.exp(-1j*phasemat)*dat,axis = (1,2))
        sus_jk = util.jackknife(sign,sus)
        
        sus_val[kid[0],kid[1]] = sus_jk[0]
        sus_err[kid[0],kid[1]] = sus_jk[1]
    
    return sus_val,sus_err
```

```python
def get_zz_nn(path):
    '''Load measurements from SHERLOCK DQMC runs and look at them'''
    Nx, Ny, bps,U, tp, nflux, mu, beta, N, L, dt ,degen_ij= \
    util.load_firstfile(path,'metadata/Nx','metadata/Ny','metadata/bps','metadata/U',\
                        'metadata/t\'','metadata/nflux','metadata/mu','metadata/beta',\
                        'params/N','params/L','params/dt','params/degen_ij')
    #print(f'Interactions:U = {U}, t = 1, tp = {tp}, mu = {mu}, bps={bps}, nflux = {nflux}')
    #print(f'Temerature: beta = {beta} dt = {dt} L = {L}')
    
    n_sample, sign, zz,nn = util.load(path,"meas_eqlt/n_sample", "meas_eqlt/sign",'meas_eqlt/zz','meas_eqlt/nn')
    
    zz = np.reshape(zz,(-1,Nx,Ny),order='F')
    nn = np.reshape(nn,(-1,Nx,Ny),order='F')
    
    zz_corr = zz[:,0,0]
    nn_corr = nn[:,0,0]
    
    #zz_corr = 1/4*(zz[:,0,1]+zz[:,0,-1]+zz[:,1,0]+zz[:,-1,0])
    #nn_corr = 1/4*(nn[:,0,1]+nn[:,0,-1]+nn[:,1,0]+nn[:,-1,0])
    
#     kpi = np.array([np.pi,np.pi]) #wave vector
#     phasemat = np.zeros((Nx,Ny),dtype=float)
#     for loc in product(range(Nx),range(Ny)):
#         phase = np.dot(kpi,np.array(loc))
#         phasemat[loc[0],loc[1]] = phase
        

    #print(np.exp(1j*phasemat).real)
    #cdw = np.sum(np.exp(1j*phasemat).real*nn,axis = (1,2))
    #sdw = np.sum(np.exp(1j*phasemat).real*zz,axis = (1,2))
    
    cdw_jk = util.jackknife(sign,nn_corr)
    sdw_jk = util.jackknife(sign,zz_corr)
    #print(sdw_jk.shape)
    
    return cdw_jk, sdw_jk
```
