```python
def eps_k(ky, kx, ty=1.0, tx=1.0, tp=0.0, typp=0.0, txpp=0.0):
    return (-2*ty*np.cos(2*np.pi*ky) - 2*tx*np.cos(2*np.pi*kx)
            - 4*tp*np.cos(2*np.pi*ky)*np.cos(2*np.pi*kx)
            - 2*typp*np.cos(4*np.pi*ky) - 2*txpp*np.cos(4*np.pi*kx))
```

```python
def Beps_k(Ny, Nx, ty=1.0, tx=1.0, tp=0.0, typp=0.0, txpp=0.0, nflux=0):
    B = nflux/(Ny*Nx)
    tij = np.zeros((Ny*Nx, Ny*Nx), dtype=np.complex)
    for iy in range(Ny):
        for ix in range(Nx):
            iy1 = (iy + 1) % Ny
            ix1 = (ix + 1) % Nx
                #jx    jy    ix    iy
            tij[ix + Nx*iy1, ix + Nx*iy] += ty * np.exp(1j*np.pi*B*( 1* ix + (0 if iy1 > iy else  Ny* ix)))
            tij[ix + Nx*iy, ix + Nx*iy1] += ty * np.exp(1j*np.pi*B*(-1* ix + (0 if iy1 > iy else -Ny* ix)))
            tij[ix1 + Nx*iy, ix + Nx*iy] += tx * np.exp(1j*np.pi*B*( 1*-iy + (0 if ix1 > ix else  Nx*-iy)))
            tij[ix + Nx*iy, ix1 + Nx*iy] += tx * np.exp(1j*np.pi*B*(-1*-iy + (0 if ix1 > ix else -Nx*-iy)))

            tij[ix1 + Nx*iy1, ix + Nx*iy] += tp * np.exp(1j*np.pi*B*( 1*ix  +  1*-iy  + (0 if iy1 > iy else  Ny*ix1) + (0 if ix1 > ix else  Nx*-iy1) + (-Nx*Ny if ix1 < ix and iy1 < iy else 0)))
            tij[ix + Nx*iy, ix1 + Nx*iy1] += tp * np.exp(1j*np.pi*B*(-1*ix1 + -1*-iy1 + (0 if iy1 > iy else -Ny*ix ) + (0 if ix1 > ix else -Nx*-iy ) + ( Nx*Ny if ix1 < ix and iy1 < iy else 0)))
            tij[ix1 + Nx*iy, ix + Nx*iy1] += tp * np.exp(1j*np.pi*B*(-1*ix  +  1*-iy1 + (0 if iy1 > iy else -Ny*ix1) + (0 if ix1 > ix else  Nx*-iy ) + ( Nx*Ny if ix1 < ix and iy1 < iy else 0)))
            tij[ix + Nx*iy1, ix1 + Nx*iy] += tp * np.exp(1j*np.pi*B*( 1*ix1 + -1*-iy  + (0 if iy1 > iy else  Ny*ix ) + (0 if ix1 > ix else -Nx*-iy1) + (-Nx*Ny if ix1 < ix and iy1 < iy else 0)))

            iy2 = (iy + 2) % Ny
            ix2 = (ix + 2) % Nx
            tij[ix + Nx*iy2, ix + Nx*iy] += typp * np.exp(1j*np.pi*B*( 2* ix + (0 if iy2 > iy else  Ny* ix)))
            tij[ix + Nx*iy, ix + Nx*iy2] += typp * np.exp(1j*np.pi*B*(-2* ix + (0 if iy2 > iy else -Ny* ix)))
            tij[ix2 + Nx*iy, ix + Nx*iy] += txpp * np.exp(1j*np.pi*B*( 2*-iy + (0 if ix2 > ix else  Nx*-iy)))
            tij[ix + Nx*iy, ix2 + Nx*iy] += txpp * np.exp(1j*np.pi*B*(-2*-iy + (0 if ix2 > ix else -Nx*-iy)))
    print(np.nonzero(np.transpose(tij[0])))
    #BEWARE
    #K[i,j] on paper is K[j,i] in code here because python is row major and 
    #C code is column major
    print(np.transpose(tij)[0,np.nonzero(tij[0])[0]])
    
#     plt.figure()
#     plt.imshow(tij.real);plt.colorbar()
#     plt.figure()
#     plt.imshow(tij.imag);plt.colorbar()
#     plt.figure()
#     plt.imshow((tij.T.conj() - tij).real);plt.colorbar()
#     plt.figure()
#     plt.imshow((tij.T.conj() - tij).imag);plt.colorbar()
    return np.linalg.eigh(tij)[0]
```

```python
#  np.save("ekkk.npy", ek)
# ek = np.load("ekkk.npy")
Nx = 10; Ny = 10;
nflux= 1
ek = Beps_k(Nx, Ny, tx=-1,ty=-1, tp=0, nflux=nflux) #, tx=4/3, ty=2/3, typp=1/6, nflux=1)
w = np.linspace(-4.35, 5.75, 3241)
#print(w)
Nw = -1/np.pi * (1/(w - ek[:, None] + 0.01j)).imag.mean(0)
plt.figure(figsize=(10,8))
plt.plot(w, Nw, lw=1)
plt.ylim(ymin=0)
#plt.xlim(-4.35, 5.75)
plt.xlabel("$\omega$")
plt.ylabel("$N(\omega)$")
plt.title(f"$B={nflux}/{Nx*Ny}$")
# plt.savefig("Nw1.png", dpi=300)
plt.show()
# Nw.sum() * (w[1] - w[0])
plt.figure(figsize=(10,8))
plt.hist(ek,bins=100,density=True,label="numerical eigenvalue histogram")
#plt.hist(eps_k_mat.flatten(),bins=80,density=True,label="formula eigenvalue hist")
#plt.plot(earr,Dearr,label="poisson proxy, formula eigenvals")
#plt.plot(earr,1/(4*np.pi)*np.ones(earr.shape),label="band edge t'=0")
plt.title(fr"2D Nx={Nx} Ny={Ny} nflux={nflux}")
plt.xlabel("energy E/t")
plt.ylabel("D(E)")
plt.legend(loc="best")
```

```python
def nnn(ek, beta):
    nk = 1.0/(np.exp(beta * ek) + 1.0)
    return 2.0*nk.mean()


def chiii(ek, beta):
    chi = np.cosh(beta*ek/2)**-2
    return beta/4 * chi.mean()


def mmmu(ek, beta, n):
    return scipy.optimize.brentq(lambda mu: nnn(ek - mu, beta) - n,
                                 ek.min() - 4.0/beta,
                                 ek.max() + 4.0/beta)

def plot_vhs(tp, tpp, ny=32, nx=32, beta=100, **kwargs):
    ek = Beps_k(ny, nx, t=1.0, tp=tp, tpp=tpp, nflux=1)
    wvhs = eps_k(0.5, 0.0, tp=tp, tpp=tpp)
    nvhs = nnn(ek - wvhs, beta)
    wmin = eps_k(0.0, 0.0, tp=tp, tpp=tpp)
    wvhs = eps_k(0.5, 0.0, tp=tp, tpp=tpp)
    nvhs = nnn(ek - wvhs, beta)

    mus = np.linspace(wmin, wmin+4, 1601)
    ns = np.array([nnn(ek - mu, beta) for mu in mus])
    chis = np.array([chiii(ek - mu, beta) for mu in mus])

    plt.plot(ns, chis, '-', lw=0.75, **kwargs)
#     plt.plot(mus-wvhs, chis, '-', lw=0.75, **kwargs)

#     plt.ylim(0, chis.max()*1.05)
#     plt.xlim(wmin, wmin+4)
#     plt.xlabel(r"$\mu$")
#     plt.ylabel(r"$\chi$")
#     plt.show()

#     plt.plot(ns, chis, '-', lw=0.5)
#     plt.plot([nvhs, nvhs], [0, 2], '--', lw=0.5)
#     plt.ylim(0, chis.max()*1.05)
#     plt.xlim(0, 1)
#     plt.xlabel(r"$\langle n \rangle$")
#     plt.ylabel(r"$\chi$")
#     plt.show()
```
