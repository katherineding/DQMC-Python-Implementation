# Equal time measurements of size Nx x Ny

```python
def plot_helper_Nx_Ny(mi,mj):
    "note all variables in here are kernel variables,\
    and depends entirely on execution order"
    colors = plt.cm.plasma(0.95*(1-nf_arr/nf_arr.max()))
    for j in range(nU):
        Us = Us_list[j]
        for i in range(nT):
            betas=betas_list[i]
            plt.figure(figsize=(8,6));
            plt.title(tt_prefix + f" beta={betas} U={Us}")
            offset = nf_arr/nf_arr.max()*3 
            n=0
            for k in range(nF):
                nflux = nflux_list[k]
                plt.errorbar(mu_ld,data_ld[i,j,k,:,mi,mj].real,\
                             yerr = dataerr_ld[i,j,k,:,mi,mj],\
                             label=f"B={nflux_list[k]}/{N}",lw=2,\
                             color=colors[k])
                n+=1
            plt.xlabel(r"chemical potential $\mu$")
            plt.ylabel(f"{meas_name}")
            plt.xlim(-6,6)
            plt.grid(True)
            plt.axvline(x=0,color='k',lw=2,ls=":")
            plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
            #plt.savefig(f'{plotroot}{Nx}x{Ny}_tp{tp}_beta{betas}_U{Us}_n_vs_mu.png',dpi=300,bbox_inches="tight")
            plt.show()
```

```python
meas_list = ['zz','xx','nn','pair_sw','g00']
for meas_name in meas_list:
    #load data
    print(meas_name)
    mu_ld = np.load(resultroot+f"{meas_name}/Nx{Nx}_Ny{Ny}_tp{tp}_mu.npy")
    data_ld = np.load(resultroot+f"{meas_name}/Nx{Nx}_Ny{Ny}_tp{tp}_{meas_name}.npy")
    dataerr_ld = np.load(resultroot+f"{meas_name}/Nx{Nx}_Ny{Ny}_tp{tp}_{meas_name}err.npy")
    num_ok = np.count_nonzero(~np.isnan(data_ld))
    print(f"{int(num_ok)}/{data_ld.size} = {int(num_ok)/data_ld.size:.3g} of data is saved in directory {resultroot}{meas_name}/")
    print(f"{meas_name}_ld.shape = {data_ld.shape}")
    #plot data
    plot_helper_Nx_Ny(0,0)
```

```python
def plot_helper_Nx_Ny_half_fill(mi,mj):
    "note all variables in here are kernel variables,\
    and depends entirely on execution order"
    colors = plt.cm.plasma(0.95*(1-nf_arr/nf_arr.max()))
    for i in range(nT-1,nT):
        betas = betas_list[i]
        plt.figure();
        plt.title(tt_prefix + f" n=1.0 beta={betas}")
        for j in range(nU):
            Us = Us_list[j]
            plt.errorbar(nf_arr,data_ld[i,j,:,nmu//2,mi,mj].real,\
                         yerr = dataerr_ld[i,j,:,nmu//2,mi,mj],\
                         label=f"U={Us}",lw=2,fmt="s:")
            #plt.figure()
            #plt.plot(nf_arr,data_ld[i,j,:,nmu//2,0,1].real,'s')
            #plt.plot(nf_arr,data_ld[i,j,:,nmu//2,0,-1].real,'^')
            #plt.plot(nf_arr,data_ld[i,j,:,nmu//2,1,0].real,"v")
            #plt.plot(nf_arr,data_ld[i,j,:,nmu//2,-1,0].real,'.')
            #print(data_ld[i,j,:,nmu//2,0,1].real,\
            #      data_ld[i,j,:,nmu//2,1,0].real,\
            #      data_ld[i,j,:,nmu//2,0,-1].real,\
             #     data_ld[i,j,:,nmu//2,-1,0].real)
                         #color=colors[k])
        plt.xlabel(r"nflux")
        #plt.xscale("log")
        plt.ylabel(f"{meas_name}[{mi},{mj}]")
        #plt.ylim([0.9,1.1])
        #plt.xlim([0.1,1.1])
        #plt.ylim([-2,0])
#         plt.xscale('log')
        #plt.ylim([0.48,0.92])
        plt.grid(True)
        #plt.axvline(x=0,color='k',lw=2,ls=":")
        plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
        #plt.savefig(f'{plotroot}{meas_name}/{Nx}x{Ny}_tp{tp}_beta{betas}_n1.0_{meas_name}_2.png',dpi=300,bbox_inches="tight")
        plt.show()
```

```python
def plot_helper_Nx_Ny_half_fill_nn():
    "note all variables in here are kernel variables,\
    and depends entirely on execution order"
    colors = plt.cm.plasma(0.95*(1-nf_arr/nf_arr.max()))
    for i in range(nT-1,nT):
        betas = betas_list[i]
        plt.figure(figsize=(5,6));
        plt.title(tt_prefix + f" n=1.0 beta={betas}")
        for j in range(nU):
            Us = Us_list[j]
            t = 1/4*(data_ld[i,j,:,nmu//2,0,1] + data_ld[i,j,:,nmu//2,0,-1] + \
                data_ld[i,j,:,nmu//2,1,0] + data_ld[i,j,:,nmu//2,-1,0])
            #TODO: is this expression for mean error correct?
            te = 1/4*np.sqrt(dataerr_ld[i,j,:,nmu//2,0,1] **2 + dataerr_ld[i,j,:,nmu//2,0,-1]**2 + \
                         dataerr_ld[i,j,:,nmu//2,-1,0]**2 + dataerr_ld[i,j,:,nmu//2,1,0]**2)
            plt.errorbar(nf_arr,t.real,yerr = te,\
                         label=f"U={Us}",lw=2,fmt="s:")
        plt.xlabel(r"nflux")
        #plt.xscale("log")
        plt.ylabel(f"{meas_name} nearest-neighbor")
        #plt.ylim([0.89,1])
        #plt.xlim([0.1,1.1])
        #plt.ylim([-2,0])
#         plt.xscale('log')
        #plt.ylim([0.48,0.92])
        plt.grid(True)
        #plt.axvline(x=0,color='k',lw=2,ls=":")
        plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
        plt.savefig(f'{plotroot}{meas_name}/{Nx}x{Ny}_tp{tp}_beta{betas}_n1.0_{meas_name}_nn.png',dpi=300,bbox_inches="tight")
        plt.show()
```

```python
meas_list = ['nn']
for meas_name in meas_list:
    #load data
    print(meas_name)
    mu_ld = np.load(resultroot+f"{meas_name}/Nx{Nx}_Ny{Ny}_tp{tp}_mu.npy")
    data_ld = np.load(resultroot+f"{meas_name}/Nx{Nx}_Ny{Ny}_tp{tp}_{meas_name}.npy")
    dataerr_ld = np.load(resultroot+f"{meas_name}/Nx{Nx}_Ny{Ny}_tp{tp}_{meas_name}err.npy")
    num_ok = np.count_nonzero(~np.isnan(data_ld))
    print(f"{int(num_ok)}/{data_ld.size} = {int(num_ok)/data_ld.size:.3g} of data is saved in directory {resultroot}{meas_name}/")
    print(f"{meas_name}_ld.shape = {data_ld.shape}")
    #plot data
    plot_helper_Nx_Ny_half_fill_nn()
```

```python
mu_ld = np.load(resultroot+f"energy/Nx{Nx}_Ny{Ny}_tp{tp}_mu.npy")
energy = np.load(resultroot+f"energy/Nx{Nx}_Ny{Ny}_tp{tp}_energy.npy")
energyerr = np.load(resultroot+f"energy/Nx{Nx}_Ny{Ny}_tp{tp}_energyerr.npy")
kinetic = np.load(resultroot+f"kinetic/Nx{Nx}_Ny{Ny}_tp{tp}_kinetic.npy")
kineticerr = np.load(resultroot+f"kinetic/Nx{Nx}_Ny{Ny}_tp{tp}_kineticerr.npy")
potential = np.load(resultroot+f"potential/Nx{Nx}_Ny{Ny}_tp{tp}_potential.npy")
potentialerr = np.load(resultroot+f"potential/Nx{Nx}_Ny{Ny}_tp{tp}_potentialerr.npy")
colors = plt.cm.plasma(0.95*(1-nf_arr/nf_arr.max()))
for j in range(nU):
    Us = Us_list[j]
    for i in range(nT):
        betas=betas_list[i]
        plt.figure(figsize=(8,6));
        plt.title(tt_prefix + f" beta={betas} U={Us}")
        n=0
        for k in range(3,4):
            nflux = nflux_list[k]
            
            plt.errorbar(mu_ld,energy[i,j,k,:].real,yerr = energyerr[i,j,k,:],fmt="-",\
                         label=f"B={nflux_list[k]}/{N} energy",lw=3)
            plt.errorbar(mu_ld,kinetic[i,j,k,:].real,yerr = kineticerr[i,j,k,:],fmt=":",\
                         label=f"B={nflux_list[k]}/{N} kinetic energy",lw=2)
            plt.errorbar(mu_ld,potential[i,j,k,:].real,yerr = potentialerr[i,j,k,:],fmt="--",\
                         label=f"B={nflux_list[k]}/{N} potential energy",lw=2)
            plt.errorbar(mu_ld,(kinetic[i,j,k,:]+potential[i,j,k,:]).real,\
                         yerr = kineticerr[i,j,k,:]+potentialerr[i,j,k,:],fmt="o",\
                         label=f"B={nflux_list[k]}/{N} sum",lw=1)
            n+=1
        plt.xlabel(r"chemical potential $\mu$")
        #plt.ylabel(f"{meas_name}")
        plt.ylim(np.nanmin(data_ld[i,j,:,:].real)-0.1,np.nanmax(data_ld[i,j,:,:].real)+0.1)
        plt.xlim(-6,6)
        plt.grid(True)
        plt.axvline(x=0,color='k',lw=2,ls=":")
        plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
        #plt.savefig(f'{plotroot}{Nx}x{Ny}_tp{tp}_beta{betas}_U{Us}_n_vs_mu.png',dpi=300,bbox_inches="tight")
        plt.show()
```
