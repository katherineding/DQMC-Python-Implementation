```python
print("nflux_list:",nflux_list)
print("betas_list:",betas_list)
print("Us_list:",Us_list)
```

# xx Conductivity

```python
sigxx_ld = np.load(resultroot+f"Nx{Nx}_Ny{Ny}_n{nt}_tp{tp}_sigxx.npy")
halfomega_ld = np.load(resultroot+f"Nx{Nx}_Ny{Ny}_n{nt}_tp{tp}_halfomega.npy")
halfdomega_ld = np.load(resultroot+f"Nx{Nx}_Ny{Ny}_n{nt}_tp{tp}_halfdomega.npy")
num_ok = np.count_nonzero(~np.isnan(sigxx_ld))
print(f"{int(num_ok)}/{sigxx_ld.size} = {int(num_ok)/sigxx_ld.size:.3g} of data is saved in directory {resultroot}")
print(f"sigxx_ld.shape = {sigxx_ld.shape} =? ({nT},{nU},{nF}, bs, nw//2) ")
print(f"halfomega_ld.shape = {halfomega_ld.shape}")
bs_xx = sigxx_ld.shape[3]
nw = sigxx_ld.shape[4]*2
```

## Zero Frequency xx conductivity bootstrap hisogram

```python
for k in range(nF):
    nflux = nflux_list[k]
    pathroot = f"/home/{user}/Documents/GradWork/Devereaux/Data-DQMC/{Nx}x{Ny}/nflux{nflux}/n{nt}/"
    _, _, mu_dict = da.readmu(pathroot,"best_mu.txt")
    for j in Uorder[1:]:
        Us = Us_list[j]
        for i in border:
            betas = betas_list[i]
            try:
                mu = mu_dict[(betas,Us)]
            except KeyError as e:
                print(e)
                continue
            path = pathroot + f"beta{betas}_U{Us}_mu{mu:.3f}/"
            try:
                plt.figure()
                plt.title(tt_prefix + f" nflux={nflux} beta={betas} U={Us} bs={bs_xx}")
                plt.hist(sigxx_ld[i,j,k,:,0],bins=20)
                plt.xlabel(r"$\sigma_{xx}(\omega=0)$ distribution")
                plt.ylabel(r"counts")
                plt.show()
            except ValueError as e:
                print(e) 
                continue
```

## Optical xx Conductivity Bootstrap

```python
for k in range(0,nF):
    nflux = nflux_list[k];
    for j in Uorder[1:]:
        U_s = Us_list[j]
        for i in border:
            beta_s = betas_list[i]
            w = halfomega_ld[i,j,k,:]
            plt.figure();
            plt.title(tt_prefix + f" nflux={nflux} beta={beta_s} U={U_s} bs = {bs_xx}")
            plt.plot(w,sigxx_ld[i,j,k,:,:].T,color='k',lw=1)
            plt.grid(True)
            plt.ylabel(r"$\sigma_{xx}(\omega)$")
            plt.xlabel(r"real frequency $\omega$")
            plt.show()
```

## Plot $\sigma_{xx}(\omega = 0)$ 

```python
colors = plt.cm.plasma(np.linspace(0,1,nF))
for j in Uorder[1:]:
    U_s = Us_list[j]
    plt.figure();
    plt.title(tt_prefix + f" U={U_s} bs={bs_xx}")
    n=0
    for k in range(nF):
        nflux = nflux_list[k];
        xx = sigxx_ld[:,j,k,:,0]
        plt.errorbar(1/beta_list[border],np.nanmean(xx,axis=1)[border],\
                     yerr = np.nanstd(xx,axis=1)[border], 
                     fmt="^:",label=f"B={nflux}/{N}")
        n+=1
    plt.ylabel(r"Re[$\sigma_{xx}(\omega = 0)$]")
    plt.xlabel(r"temperature $T/t$")
    plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
    plt.savefig(f'{plotroot}{Nx}x{Ny}_tp{tp}_n{nt}_U{U_s}_condxx.png',dpi=300,bbox_inches="tight")
```

## $\rho_{xx}(B)|_{\omega=0}$ as function of $B$ field. Not included xy contribution yet.

```python
for i in border:
    beta_s = betas_list[i]
    plt.figure();
    plt.title(tt_prefix +fr" $\beta$={beta_s} bs={bs_xx}")
    for j in Uorder[1:]:
        U_s = Us_list[j]
        xx = sigxx_ld[i,j,:,:,0]
        mxx = np.nanmean(xx,axis=1)
        exx = np.nanstd(xx,axis=1)
        rxx = exx/mxx #relative error
        #print(np.nanstd(1/xx,axis=1),rxx/mxx)
        #plt.errorbar(B_arr,1/mxx,yerr = np.nanstd(1/xx,axis=1), fmt="^:",label=f"U={U_s} 1")
        plt.errorbar(B_arr,1/mxx,yerr = rxx*1/mxx, fmt="s:",label=f"U={U_s}")
    plt.xlabel(r"B [$\Phi_0/a^2$]")
    plt.ylabel(r"$\rho_{xx}(\omega = 0)$  [$\hbar/e^2$]")
    #plt.ylim(0,10)
    #plt.xlim(-0.01,0.25)
    plt.grid(True)
    plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
    plt.savefig(f'{plotroot}{Nx}x{Ny}_tp{tp}_n{nt}_beta{beta_s}_rhoxx.png',dpi=300,bbox_inches="tight")
```

## Magnetoresistance as function of $B$ field. Not included xy contribution yet.

```python
for i in border[2:]:
    beta_s = betas_list[i]
    plt.figure();
    plt.title(tt_prefix +fr" $\beta$={beta_s} bs={bs_xx}")
    for j in Uorder[1:]:
        U_s = Us_list[j]
        sxx = sigxx_ld[i,j,:,:,0]
        mxx = np.nanmean(sxx,axis=1)
        #zero field
        s0 = sxx[0,:]
        m0 = np.nanmean(s0)
        re0 = np.nanstd(s0)/m0
        #some field
        sB = sxx[1:,:]
        mB = np.nanmean(sB,axis=1)
        reB = np.nanstd(sB,axis=1)/mB
        #print("finite field errors",reB)
        #print("zero frequency errors",re0)
        retotal = np.sqrt(np.power(reB,2)+np.power(re0,2))
        #print("total relative errors",retotal,retotal.shape)
        retotal = np.insert(retotal,0,re0)
        #print("total relative errors",retotal,retotal.shape,"\n")
        MRB = s0/sxx - 1
        MRm = np.nanmean(MRB,axis=1)
        plt.errorbar(B_arr ,MRm, yerr = retotal*(mxx/m0), fmt="^--",label=f"U={U_s}")
        plt.axhline(y=0,lw=1,color='k')
    plt.xlabel(r"B [$\Phi_0/a^2$]")
    plt.ylabel("MR(B)")
    #plt.grid(True)
    plt.ylim(-0.05,1)
    plt.xlim(-0.01,0.25)
    plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
    plt.savefig(f'{plotroot}{Nx}x{Ny}_tp{tp}_n{nt}_beta{beta_s}_MR.png',dpi=300,bbox_inches="tight")
```
