```python
Nx = "4"; Ny = "4"; nt = "0.125"; tp='0'; N = int(Nx) * int(Ny)
if nt == "1.0":
    nflux_list = ['0','1','4','8','12','16'];
else:
    nflux_list = ['4'];
nf_arr = np.array(list(map(float,nflux_list)))
nF = len(nflux_list)
tt_prefix = f"{Nx}x{Ny} t'={tp} n={nt}"
```

```python
for k in range(nF):
    nflux = nflux_list[k]
    pathroot = f"/home/{user}/Documents/GradWork/Devereaux/Data-DQMC/{Nx}x{Ny}/nflux{nflux}/n{nt}/"
    plt.figure();
    plt.title(tt_prefix + f" nflux={nflux_list[k]}")
    betas_list, Us_list, mu_dict = da.readmu(pathroot,"best_mu.txt",show=True)
    plt.show()
    
if nt == "0.95" or nt == "1.0":
    pathroot = f"/home/{user}/Documents/GradWork/Devereaux/Data-DQMC/{Nx}x{Ny}/nflux1/n{nt}/"      
else:
    pathroot = f"/home/{user}/Documents/GradWork/Devereaux/Data-DQMC/{Nx}x{Ny}/nflux4/n{nt}/"  
betas_list, Us_list, mu_dict = da.readmu(pathroot,"best_mu.txt",show=False)
nT = len(betas_list); nU = len(Us_list)
beta_list = np.array(list(map(float,betas_list)))
U_list = np.array(list(map(float,Us_list)))
border = np.argsort(beta_list); Uorder = np.argsort(U_list)
```

```python
nw = 400
print("nflux_list:",nflux_list)
print("betas_list:",betas_list)
print("Us_list:",Us_list)
bs = 1
```

```python
dos_collect = np.full((nT,nU,nF,bs,nw),np.nan)
dos_omega_collect = np.full((nT,nU,nF,nw),np.nan)
dos_domega_collect = np.full((nT,nU,nF,nw),np.nan)

for k in range(nF):
    nflux = nflux_list[k]
    pathroot = f"/home/{user}/Documents/GradWork/Devereaux/Data-DQMC/{Nx}x{Ny}/nflux{nflux}/n{nt}/"
    _, _, mu_dict = da.readmu(pathroot,"best_mu.txt")
    for j in Uorder[2:]:
        Us = Us_list[j]
        for i in border[1:]:
            betas = betas_list[i]
            try:
                mu = mu_dict[(betas,Us)]
            except KeyError as e:
                print(e);
                continue
            path = pathroot + f"beta{betas}_U{Us}_mu{mu:.3f}/"
            print(path)
            _, ds, w, dw = dos.dos(path,nw=nw,bs=bs,method = "BT",anneal_arr = None,checks=True)
            dos_collect[i,j,k,:,:] = ds
            dos_omega_collect[i,j,k] = w
            dos_domega_collect[i,j,k] = dw
            
```

```python
np.save(resultroot+f"Nx{Nx}_Ny{Ny}_n{nt}_tp{tp}_dos.npy",dos_collect)
np.save(resultroot+f"Nx{Nx}_Ny{Ny}_n{nt}_tp{tp}_dosomega.npy",dos_omega_collect)
np.save(resultroot+f"Nx{Nx}_Ny{Ny}_n{nt}_tp{tp}_dosdomega.npy",dos_domega_collect)
```
