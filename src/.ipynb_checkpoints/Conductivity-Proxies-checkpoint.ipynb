```python
def check_div_vals(path):
    U, beta, Nx, Ny, bps, nflux, N, L, dt = util.load_firstfile(path,"metadata/U","metadata/beta",\
            "metadata/Nx","metadata/Ny","metadata/bps","metadata/nflux", "params/N","params/L","params/dt")    
    ns, s, jj = util.load(path, "meas_uneqlt/n_sample", "meas_uneqlt/sign", "meas_uneqlt/jj")
    #reshape jj into more intuitive shape
    #jj.shape = -1, L, bps, bps, Ny, Nx
    jj = np.reshape(jj,(-1,Nx,Ny,bps,bps,L),order='F')
    jj = np.transpose(jj,(0,5,4,3,1,2))
    
    #filling and sign info
    if da.info(path,uneqlt=False,show=False) != 0: 
        return None
    
    #use only completed bins
    mask = ns == ns.max(); nbin = mask.sum()
    ns, s, jj = ns[mask], s[mask], jj[mask]

    jjq0 = jj.sum((-1,-2))   #take q == 0
    #forumlas specific to tp = 0, bps=2 case
    jxjxq0 = jjq0[:,:, 0, 0]; jyjyq0 = jjq0[:,:, 1, 1];
    jxjyq0 = jjq0[:,:, 1, 0]; jyjxq0 = jjq0[:,:, 0, 1];

    # <(jx + i*jy) (jx-i*jy)> = <jxjx> + <jyjy> - 2i<jxjy> 
    # FIT object: chi''_xx -i chi''_xy
    #divide each bin by average sign
    chiq0_symm  = -0.5*(jxjxq0 + jyjyq0)/np.mean(s.real)
    chiq0_asymm = 0.5*1j*(jxjyq0 - jyjxq0)/np.mean(s.real)
    #Maxent can only work with real values
    chiq0_symm  = chiq0_symm.real
    chiq0_asymm  = chiq0_asymm.real
    chiq0 = chiq0_symm + chiq0_asymm #without data at tau = beta
    
    taus = np.arange(L)*dt; treq = np.linspace(0,taus[-1],100)
    spl_symm = InterpolatedUnivariateSpline(taus,chiq0_symm.mean(0),k=3,ext=2,check_finite=True)
    spl_asymm = InterpolatedUnivariateSpline(taus,chiq0_asymm.mean(0),k=3,ext=2,check_finite=True)
    spl_all = InterpolatedUnivariateSpline(taus,chiq0.mean(0),k=3,ext=2,check_finite=True)
    for n in range(4):
        v1 = spl_symm.derivative(n);
        v2 = spl_asymm.derivative(n);
        v3 = spl_all.derivative(n);
        print(f"deriv order {n} \tsym {v1(beta/2):.3e} + asym {v2(beta/2):.3e} = all {v3(beta/2):.3e}")
        plt.figure()
        plt.title(f"derivative {n} U ={U} beta={beta}")
        plt.plot(treq,v1(treq),'.',label="symm")
        plt.plot(treq,v2(treq),'.',label="asymm")
        #plt.plot(treq,v3(treq),'.',label="all")
        plt.axhline(y=0,color='k',lw=1)
        plt.axvline(x=beta/2,color="k",lw=1)
        plt.legend(loc="best")
        plt.xlim(-beta/L,beta+beta/L)
        plt.show()
    return None
```

## Proposed proxy

```python
def proposed_proxy(path, bs_rep = 0):
    U, beta, Nx, Ny, bps, nflux, N, L, dt = util.load_firstfile(path,"metadata/U","metadata/beta",\
            "metadata/Nx","metadata/Ny","metadata/bps","metadata/nflux", "params/N","params/L","params/dt")    
    ns, s, jj = util.load(path, "meas_uneqlt/n_sample", "meas_uneqlt/sign", "meas_uneqlt/jj")
    #reshape jj into more intuitive shape
    #jj.shape = -1, L, bps, bps, Ny, Nx
    jj = np.reshape(jj,(-1,Nx,Ny,bps,bps,L),order='F')
    jj = np.transpose(jj,(0,5,4,3,1,2))
    
    #filling and sign info
    if da.info(path,uneqlt=False,show=False) > 0: 
        return np.nan, np.nan, np.nan, np.nan, np.nan, np.nan,np.nan,np.nan
    
    #use only completed bins
    mask = ns == ns.max(); nbin = mask.sum()
    ns, s, jj = ns[mask], s[mask], jj[mask]

    jjq0 = jj.sum((-1,-2))   #take q == 0
    #forumlas specific to tp = 0, bps=2 case
    jxjxq0 = jjq0[:,:, 0, 0]; jyjyq0 = jjq0[:,:, 1, 1];
    jxjyq0 = jjq0[:,:, 1, 0]; jyjxq0 = jjq0[:,:, 0, 1];

    # <(jx + i*jy) (jx-i*jy)> = <jxjx> + <jyjy> - 2i<jxjy> 
    # FIT object: chi''_xx -i chi''_xy
    #divide each bin by average sign and take real part
    chiq0_symm  = (-0.5*(jxjxq0 + jyjyq0)/np.mean(s)).real
    chiq0_asymm = (0.5*1j*(jxjyq0 - jyjxq0)/np.mean(s)).real
    chiq0 = chiq0_symm + chiq0_asymm #without data at tau = beta
    
    taus = np.arange(L)*dt; treq = np.linspace(0,taus[-1]-dt,100)
    
    f = lambda x , p1 , p3, p5 : (x-beta/2)*(p1 +(x-beta/2)**2*(p3 + (x-beta/2)**2 * p5))
    #if bs_rep > 0, bootstrap resample that many times
    if bs_rep > 0:
        print(f"bootstrap # of reps: {bs_rep}")
        ps = np.empty(bs_rep)*np.nan;
        pa = np.empty(bs_rep)*np.nan
        pextra = np.empty(bs_rep)*np.nan
        for i in range(bs_rep):
            resample = np.random.randint(nbin, size=nbin)
            spl_symm = InterpolatedUnivariateSpline(taus,chiq0_symm[resample].mean(0),k=3,ext=2,check_finite=True)
            spl_asymm = InterpolatedUnivariateSpline(taus,chiq0_asymm[resample].mean(0),k=3,ext=2,check_finite=True)
            spl_all = InterpolatedUnivariateSpline(taus,chiq0[resample].mean(0),k=3,ext=2,check_finite=True)
            sdiv2 = spl_symm.derivative(2);
            adiv3 = spl_asymm.derivative(3);  
            adiv1 = spl_asymm.derivative(1);
            #proxy for rho_xx
            ps[i] = sdiv2(beta/2)/(2*np.pi*spl_all(beta/2)**2)
            pextra[i] = adiv1(beta/2)/(spl_symm(beta/2)**2*beta**1)
            #proxy for sigma_xy
        
        return np.mean(chiq0_asymm[:,0]), np.std(chiq0_asymm[:,0]), np.mean(ps), np.std(ps), np.nan, np.nan, np.mean(pextra),np.std(pextra)

    #no bootstrap
    spl_symm = InterpolatedUnivariateSpline(taus,chiq0_symm.mean(0),k=3,ext=2,check_finite=True)
    spl_asymm = InterpolatedUnivariateSpline(taus,chiq0_asymm.mean(0),k=3,ext=2,check_finite=True)
    spl_all = InterpolatedUnivariateSpline(taus,chiq0.mean(0),k=3,ext=2,check_finite=True)
    sdiv2 = spl_symm.derivative(2); #print(sdiv2(beta/2))
    adiv3 = spl_asymm.derivative(3); #print(adiv3(beta/2))
    adiv1 = spl_asymm.derivative(1); #print(adiv1(beta/2))
    ps = sdiv2(beta/2)/(2*np.pi*spl_all(beta/2)**2*beta**2)
    pa = np.nan
    pextra = adiv1(beta/2)/(spl_symm(beta/2)**2)
    print(f"xy 1st div at tau = 0: {adiv1(0):.3g}")
    print(f"xy 1st div at tau = beta/2: {adiv1(beta/2):.3g}")
    print(f"xy intercept at tau = 0: {spl_asymm(0):.3g}")
    v = 0.5*adiv1(beta/2)/spl_asymm(0)
    if v > 1:
        print(f"BAD div1(beta/2)/(2*intercept) = {0.5*adiv1(beta/2)/spl_asymm(0):.3g}")
    elif 0.5*adiv1(beta/2)/spl_asymm(0) < 0:
        print(f"OK div1(beta/2)/(2*intercept) = {0.5*adiv1(beta/2)/spl_asymm(0):.3g}")
        try:
            sol = root_scalar(F,args=(v,beta),x0=1,bracket=[1e-3,4])
            print(sol)
            pa = (sol.root)**2 * spl_asymm(0)
        except:
            print("fail")
    elif 0.5*adiv1(beta/2)/spl_asymm(0) > 0:
        print(f"BAD div1(beta/2)/(2*intercept) = {0.5*adiv1(beta/2)/spl_asymm(0):.3g}")

    #root_scalar(F,args=(,beta))
    #Ca = np.cov(chiq0_asymm,rowvar=False)
    #if np.linalg.cond(Ca) > 1e10:
    #    print(f"covariance matrix condition number {np.linalg.cond(Ca)}")
    #    return np.nan, np.nan, np.nan, np.nan, np.nan, np.nan
    #popt, pcov = curve_fit(f,taus,np.mean(chiq0_asymm,axis=0),p0=[0,0,0],sigma=Ca)
    #perr = np.sqrt(np.diag(pcov))
    #s_xy = (adiv1(beta/2)**2)/adiv3(beta/2) + np.mean(chiq0_asymm[:,0])*beta
    #RH = s_xy

    return np.mean(chiq0_asymm[:,0]), np.std(chiq0_asymm[:,0]), ps, 0, pa,0 ,pextra, 0
```

```python
def F(ti,val,beta):
    f = lambda w: -1/np.pi/np.sinh(beta*w/2)*w**3*ti/(ti**2+w**2)**2
    y,err = integrate.quad(f,0,20)
    print(y*2-val)
    return y*2-val
```

```python
def A_fit(path):
    U, beta, Nx, Ny, bps, nflux, N, L, dt = util.load_firstfile(path,"metadata/U","metadata/beta",\
            "metadata/Nx","metadata/Ny","metadata/bps","metadata/nflux", "params/N","params/L","params/dt")    
    ns, s, jj = util.load(path, "meas_uneqlt/n_sample", "meas_uneqlt/sign", "meas_uneqlt/jj")
    #reshape jj into more intuitive shape
    #jj.shape = -1, L, bps, bps, Ny, Nx
    jj = np.reshape(jj,(-1,Nx,Ny,bps,bps,L),order='F')
    jj = np.transpose(jj,(0,5,4,3,1,2))
    
    #filling and sign info
    if da.info(path,uneqlt=False,show=True) != 0: 
        return None
    
    #use only completed bins
    mask = ns == ns.max(); nbin = mask.sum()
    ns, s, jj = ns[mask], s[mask], jj[mask]

    jjq0 = jj.sum((-1,-2))   #take q == 0
    #forumlas specific to tp = 0, bps=2 case
    jxjyq0 = jjq0[:,:, 1, 0]; jyjxq0 = jjq0[:,:, 0, 1];

    # <(jx + i*jy) (jx-i*jy)> = <jxjx> + <jyjy> - 2i<jxjy> 
    # FIT object: chi''_xx -i chi''_xy
    # divide each bin by average sign
    chiq0_asymm = 0.5*1j*(jxjyq0 - jyjxq0)/np.mean(s.real)
    #Maxent can only work with real values
    chiq0_asymm  = chiq0_asymm.real
    #---------------------------------------------------------------------------------
    v_avg = np.mean(chiq0_asymm,axis=0); #correlator choice
    v_avg_mid = v_avg[L//2]; #value at beta/2
    omega = np.arange(-60,60,0.01)
    vspl = InterpolatedUnivariateSpline(np.arange(L)*dt,v_avg,k=3,ext=2,check_finite=True)
    
    taus = np.arange(0,beta-dt,0.01)
    integrals = np.empty(taus.shape[0])*np.nan
    ratios = np.empty(taus.shape[0])*np.nan
    for i in range(taus.shape[0]):
        k = kernel_s(beta,omega,taus[i])
        kspl = InterpolatedUnivariateSpline(omega,k,k=3,ext=2,check_finite=True)
        integrals[i]= kspl.integral(omega[0],omega[-1])
        try: 
            ratios[i] = vspl(taus[i])/integrals[i]
        except RuntimeError:
            ratios[i] = np.nan
    return ratios,None
```

```python
def kernel(beta,omega,tau):
    tauv = np.reshape(tau,(tau.shape[0],1))
    top = np.power(omega,2)*np.sinh(omega*(tauv-beta/2))
    bot = np.sinh(omega*beta/2)
    return top/bot
```

```python
def kernel_s(beta,omega,tau):
    #print( np.isscalar(beta),np.isscalar(tau))
    top = np.power(omega,2)*np.sinh(omega*(tau-beta/2))
    bot = np.sinh(omega*beta/2)
    return top/bot
```
