```python
Nx = "6"; Ny = "6"; nt = "0.85"; tp='0'; N = int(Nx) * int(Ny)
if nt == "1.0":
    nflux_list = ['0','1','4','8','12','16'];
else:
    nflux_list = ['0','1','2','3','4','8','12','16'];
nf_arr = np.array(list(map(float,nflux_list)))
nF = len(nflux_list)
B_arr = nf_arr/N
tt_prefix = f"{Nx}x{Ny} t'={tp} n={nt}"
```

```python
for k in range(nF):
    nflux = nflux_list[k]
    pathroot = f"/home/katherine/Documents/GradWork/Devereaux/Data-DQMC/{Nx}x{Ny}/nflux{nflux}/n{nt}/"
    plt.figure();
    plt.title(tt_prefix + f" nflux={nflux_list[k]}")
    betas_list, Us_list, mu_dict = da.readmu(pathroot,"best_mu.txt",show=True)
    plt.show()
    
if nt == "0.95" or nt == "1.0":
    pathroot = f"/home/katherine/Documents/GradWork/Devereaux/Data-DQMC/{Nx}x{Ny}/nflux1/n{nt}/"      
else:
    pathroot = f"/home/katherine/Documents/GradWork/Devereaux/Data-DQMC/{Nx}x{Ny}/nflux2/n{nt}/"  
betas_list, Us_list, mu_dict = da.readmu(pathroot,"best_mu.txt",show=False)
nT = len(betas_list); nU = len(Us_list)
beta_list = np.array(list(map(float,betas_list)))
U_list = np.array(list(map(float,Us_list)))
border = np.argsort(beta_list); Uorder = np.argsort(U_list)
```

```python
colors = plt.cm.viridis(1-nf_arr[:-1]/12)
offset = nf_arr/25
for j in Uorder[1:-1]:
    U_s = Us_list[j]
    for i in border[6:]:
        beta_s = betas_list[i]
        n=0
        plt.figure(figsize=(5,6))
        plt.title(tt_prefix + f" beta={beta_s} U={U_s} bs = {bs}")
        for k in range(nF-1):
            nflux=nflux_list[k]
            w = omega_ld[i,j,k]
            plt.plot(w,np.nanmean(dos_ld[i,j,k,:,:],axis=0)+offset[k],lw=2,color=colors[n],label=f"nflux={nflux}")
            n+=1;
        #plt.grid(True)
        #plt.ylabel(r"dos$(\omega)$")
        plt.xlabel(r"$\omega/t$")
        plt.xlim(-12,12)
        plt.ylim(-0.02,1.05)
        plt.axvline(x=0,color='k',lw=2,ls=":")
        #plt.gca().axes.get_yaxis().set_visible(False)
        #plt.legend(loc="best")
        plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
        #plt.show()
        plt.savefig(f'{plotroot}{Nx}x{Ny}_tp{tp}_n{nt}_beta{beta_s}_U{U_s}_localDOS.png',dpi=300,bbox_inches="tight")
```
