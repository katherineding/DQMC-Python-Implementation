```python
def gen_grid(nw, x_min, x_max, w_x):
    """
    generate grid with nw points scaled by the function w_x.
    want nw points
    first generate 2*nw+1 x-points, from which get 2*nw+1 w-points
    pickout odd indexed w-points as w grid
    

    w[i] = w_x((i+0.5)/nw * (x_max-x_min) + x_min)
    dw[i] = w_x[(i+1)/nw * (x_max-x_min) + x_min] - w_x[i/nw * (x_max-x_min) + x_min]

    returns w, dw
    """
    x_all = np.linspace(x_min, x_max, 2*nw+1)
    print(x_all)
    w_all = np.apply_along_axis(w_x, 0, x_all)
    print(w_all)
    return w_all[1:-1:2], np.abs(np.diff(w_all[0::2]))
```

```python
w,dw  = gen_grid(4,-1,1,lambda x: 0.4*np.sinh(2.5*x))
print(w,dw)
```
