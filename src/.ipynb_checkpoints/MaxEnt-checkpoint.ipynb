# Get $\rho_{xx}(\omega)$ as function of nflux, temp and U

```python
print("nflux_list:",nflux_list)
print("betas_list:",betas_list)
print("Us_list:",Us_list)
bs = 20
```

```python
nw = 100
sigxx_collect = np.full((nT,nU,nF,bs,nw//2),np.nan)
halfomega_collect = np.full((nT,nU,nF,nw//2),np.nan)
halfdomega_collect = np.full((nT,nU,nF,nw//2),np.nan)
w,dw = maxent.gen_grid(nw//2, 0, 1.7, lambda x: 0.4*np.sinh(2.5*x))
alpha_arr = np.logspace(0, 9, 1+10*(9-0))

for k in range(nF):
    nflux = nflux_list[k]
    pathroot = f"/home/{user}/Documents/GradWork/Devereaux/Data-DQMC/{Nx}x{Ny}/nflux{nflux}/n{nt}/"
    _, _, mu_dict = da.readmu(pathroot,"best_mu.txt")
    for j in Uorder[1:]:
        Us = Us_list[j]
        model=None
        for i in border:
            betas = betas_list[i]
            try:
                mu = mu_dict[(betas,Us)]
            except KeyError as e:
                print(e);
                continue
            path = pathroot + f"beta{betas}_U{Us}_mu{mu:.3f}/"
            if model is None:
                As, s = oc.sigma_xx(path, w,dw,bs=bs,method = "BT",checks=False,anneal_arr = None)
            else:
                As, s = oc.sigma_xx(path, w,dw,bs=bs,method = "classic",checks=False,anneal_arr = model)
            if np.all(np.isfinite(As)): model = As
            else: model = None
            #print(s)
            sigxx_collect[i,j,k,:,:] = s
            #print(sigxx_collect[i,j,k,:,:])
            halfomega_collect[i,j,k,:] = w;
            halfdomega_collect[i,j,k,:] = dw;
```

# Get $\chi^2$ from AC

```python
from scipy.interpolate import InterpolatedUnivariateSpline
def kramers_kronig(F,w,dw):
    """F = i\chi^2_{xy}(\omega) array"""
    kernel = 1/np.power(w,2)
    spl = InterpolatedUnivariateSpline(w,kernel*F*dw)
    I = spl.integral(w.min(),w.max())
    #if omega range [-infty,infty]
    if w.min() < 0:
        return I
    #if omega range[0,infty]
    else:
        return 2*I
```

```python
np.set_printoptions(precision=6)
sigxx2_collect = np.full((nT,nU,nF,bs,nw//2),np.nan, dtype=float)
chi2_collect   = np.full((nT,nU,nF,bs,nw),np.nan, dtype=float)
omega_collect  = np.full((nT,nU,nF,nw),np.nan, dtype=float)
domega_collect = np.full((nT,nU,nF,nw),np.nan, dtype=float)
sigxy_collect  = np.full((nT,nU,nF,bs),np.nan, dtype=float)
for k in range(1,nF):
    nflux = nflux_list[k]
    pathroot = f"/home/{user}/Documents/GradWork/Devereaux/Data-DQMC/{Nx}x{Ny}/nflux{nflux}/n{nt}/"
    _, _, mu_dict = da.readmu(pathroot,"best_mu.txt")
    for j in Uorder[1:]:
        U_s = Us_list[j]
        model = None
        for i in border:
            beta_s = betas_list[i]
            try:
                mu = f"{mu_dict[(beta_s,U_s)]:.3f}"
            except KeyError as e:
                print("no",e,"combination");
                continue
            path= pathroot + f"beta{beta_s}_U{U_s}_mu{mu}/"
            #As, s, chi2,omega,domega = hall.get_chi2_fromsymmetry(dpath,nw=nw,bs=bs,method = "BT", checks=False,anneal_arr= model)
            if model is None:
                As, s, chi2, w, dw = hall.get_chi2_fromsymmetry(path,nw=nw,bs=bs,method = "BT",checks=False,anneal_arr = model)
            else:
                As, s, chi2, w, dw = hall.get_chi2_fromsymmetry(path,nw=nw,bs=bs,method = "classic",checks=False,anneal_arr = model)
            if np.all(np.isfinite(As)): model = As
            else: model = None
            sigxx2_collect[i,j,k,:] = s
            chi2_collect[i,j,k,:] = chi2
            omega_collect[i,j,:] = w
            domega_collect[i,j,:] = dw 
            plt.figure()
            plt.title(tt_prefix+fr" nflux={nflux} U={U_s} $\beta$={beta_s}")
            plt.plot(w,(chi2/w**2).T,lw=1)
            plt.xlim(-10,10)
            plt.show()
            for l in range(bs):
                sigxy_collect[i,j,k,l] = kramers_kronig(chi2[l,:],w,dw)
                #print(kramers_kronig(chi2[l,:],w,dw))
```

```python
np.save(resultroot+f"Nx{Nx}_Ny{Ny}_n{nt}_tp{tp}_sigxx2.npy",sigxx2_collect)
np.save(resultroot+f"Nx{Nx}_Ny{Ny}_n{nt}_tp{tp}_chi2.npy",chi2_collect)
np.save(resultroot+f"Nx{Nx}_Ny{Ny}_n{nt}_tp{tp}_omega.npy",omega_collect)
np.save(resultroot+f"Nx{Nx}_Ny{Ny}_n{nt}_tp{tp}_domega.npy",domega_collect)
np.save(resultroot+f"Nx{Nx}_Ny{Ny}_n{nt}_tp{tp}_sigxy.npy",sigxy_collect)
```
