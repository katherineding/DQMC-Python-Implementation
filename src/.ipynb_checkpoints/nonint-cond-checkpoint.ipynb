```python
Nx=12; Ny=12; t=1; tp=0; N = Nx*Ny; nflux = 12; alpha = 1/2
tt_prefix = f"{Nx}x{Ny} t'={tp} nflux={nflux} "
'''Diagonalize H-mat numerically: ED'''
#Hmat = tb.H_periodic_2D(Nx,Ny,t,tp, B = nflux,alpha = alpha)
Hmat = tb.H_periodic_2D_EDWIN(Ny,Nx,ty = -t,tx = -t, tp = -tp, nflux=nflux)
#plt.imshow(Hmat.real)

evals, V = np.linalg.eigh(Hmat)
print(evals)
```

```python
'''Known exact eigenvalues with no B field, should agree with ED when B = 0'''
earr = np.linspace(-5,5,4001); sig=1/100;
eps_k_mat = tb.eps_k_2D(Nx,Ny,t,tp);
'''poisson kernels'''
Dearr = np.zeros(earr.shape,dtype=float)
Dearr_ED = np.zeros(earr.shape,dtype=float)
for i in range(int(Nx)):
    for j in range(int(Ny)):
        Dearr += tb.poisson(sig,earr,eps_k_mat[i,j])/N
        Dearr_ED += tb.poisson(sig,earr,evals[j*Nx + i])/N

plt.figure()
#plt.hist(evals,bins=100,density=True,label="numerical eigenvalue histogram")
#plt.hist(eps_k_mat.flatten(),bins=100,density=True,alpha = 0.5,label="formula eigenvalue hist")
#plt.plot(earr,Dearr,label="poisson proxy, formula eigenvals B = 0")
plt.plot(earr,Dearr_ED,label="poisson proxy, numerical eigenvals")
#plt.plot(earr,1/(4*np.pi)*np.ones(earr.shape),label="band edge t'=0")
plt.title(fr"2D Nx={Nx} Ny={Ny} t'={tp} nflux={nflux}")
plt.xlabel("energy E/t")
plt.ylabel("D(E)")
plt.legend(loc="best")
#plt.xlim(-0.5,0.5)
#print(np.sort(eps_k_mat.flatten()))
```

```python
nf_arr = np.arange(1,N//2,5);nt = 0.8
omega = np.linspace(1e-3,0.5,500)
Chi_JXJX_R = np.empty((nf_arr.shape[0],omega.shape[0]),dtype=complex)
Chi_JXJY_R = np.empty((nf_arr.shape[0],omega.shape[0]),dtype=complex)
sigma_xx = np.empty((nf_arr.shape[0],omega.shape[0]),dtype=complex)
sigma_xy = np.empty((nf_arr.shape[0],omega.shape[0]),dtype=complex)
for n in range(nf_arr.shape[0]):
    nflux = nf_arr[n]
    tt_prefix = f"{Nx}x{Ny} t'={tp} beta={beta} "
    '''Diagonalize H-mat numerically: ED'''
    #Hmat = tb.H_periodic_2D(Nx,Ny,t,tp, B = nflux,alpha = alpha)
    Hmat = tb.H_periodic_2D_EDWIN(Ny,Nx,ty = -t,tx = -t, tp = -tp, nflux=nflux)
    evals, V = np.linalg.eigh(Hmat); 
    assert np.linalg.norm((V * evals) @ (V.T.conj()) - Hmat) < 1e-10
    mu_arr = np.linspace(-2,1,200); 
    n_arr = np.sum(tb.f_dist_mat(evals,mu_arr,beta),axis=0)/N*2
    nfun = interp1d(mu_arr,n_arr-nt,kind="linear",assume_sorted=True)
#     plt.figure(figsize=(10,8))
#     plt.plot(mu_arr,n_arr,label = "$n(\mu)$")
#     plt.title(fr"2D Nx={Nx} Ny={Ny} t'={tp} $\beta$={beta} nflux={nflux}")
#     plt.xlabel(r"chemical potential $\mu/t$")
#     plt.legend(loc="best")
#     plt.grid(True)
#     plt.show()
    idx = (np.abs(n_arr - nt)).argmin()
    mu_opt = brentq(nfun,mu_arr[1],mu_arr[-2])
    print(mu_opt,mu_arr[idx])
    
    Jx = tb.JX_periodic_2D(Ny,Nx,ty = -t,tx = -t, tp = -tp, nflux=nflux)
    Jy = tb.JY_periodic_2D(Ny,Nx,ty = -t,tx = -t, tp = -tp, nflux=nflux)
    Jx_sparse = csc_matrix(Jx)
    Jy_sparse = csc_matrix(Jy)

    Jx_ubasis = V.T.conj() @ Jx_sparse.dot(V)
    Jy_ubasis = V.T.conj() @ Jy_sparse.dot(V)
    JxTJx = Jx_ubasis.T *  Jx_ubasis ; #JyTJy = Jx_ubasis.T *  Jx_ubasis
    JxTJy = Jx_ubasis.T *  Jy_ubasis ; #JyTJx = Jy_ubasis.T *  Jx_ubasis
    Chi_JXJX_R[n] = Chi_R(JxTJx, omega, mu_opt, eta)
    Chi_JXJY_R[n] = Chi_R(JxTJy, omega, mu_opt, eta)
    
    sigma_xx[n] = Chi_JXJX_R[n]/(-1j*omega*N)
    sigma_xy[n] = Chi_JXJY_R[n]/(-1j*omega*N)
```

```python
plt.figure(figsize=(12,10))
for n in range(nf_arr.shape[0]):
    nflux = nf_arr[n]
    plt.plot(omega,sigma_xy[n].real,label=f"B={nflux}/{N}")
    plt.legend(loc="best")
```
