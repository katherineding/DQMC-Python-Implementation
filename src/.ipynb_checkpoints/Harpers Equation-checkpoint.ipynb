```python
def vec_in_lattice(Nx,Ny,vec):
    assert vec.shape[0] == Nx * Ny
    return np.reshape(vec,(Nx,Ny),order='F')
```

```python
Nx=16; Ny=16; t=1; tp=0; N = Nx*Ny; nflux = 1; alpha = 1/2
tt_prefix = f"{Nx}x{Ny} t'={tp} nflux={nflux} "
'''Diagonalize H-mat numerically: ED'''
#Hmat = tb.H_periodic_2D(Nx,Ny,t,tp, B = nflux,alpha = alpha)
Hmat = tb.H_periodic_2D_EDWIN(Ny,Nx,ty = -t,tx = -t, tp = -tp, nflux=nflux)
#Hmat = tb.H_periodic_2D(Nx,Ny,t,tp, B = nflux,alpha = alpha)
evals, V = np.linalg.eigh(Hmat); 
assert np.linalg.norm((V * evals) @ (V.T.conj()) - Hmat) < 1e-10
# plt.figure()
# plt.imshow(Hmat.real); plt.colorbar()
# plt.figure()
# plt.imshow(Hmat.imag); plt.colorbar()
```

```python
'''Known exact eigenvalues with no B field, should agree with ED when B = 0'''
earr = np.linspace(-6,6,4001); sig=1/100;
eps_k_mat = tb.eps_k_2D(Nx,Ny,t,tp);
'''poisson kernels'''
Dearr = np.zeros(earr.shape,dtype=float)
Dearr_ED = np.zeros(earr.shape,dtype=float)
for i in range(int(Nx)):
    for j in range(int(Ny)):
        Dearr += tb.poisson(sig,earr,eps_k_mat[i,j])/N
        Dearr_ED += tb.poisson(sig,earr,evals[j*Nx + i])/N

plt.figure(figsize=(10,8))
plt.hist(evals,bins=100,density=True,label="numerical eigenvalue histogram")
#plt.plot(earr,Dearr_ED,label="poisson proxy, numerical DOS")
#plt.plot(earr,1/(4*np.pi)*np.ones(earr.shape),label="band edge t'=0")
plt.title(tt_prefix + " DOS")
plt.xlabel("energy E/t")
plt.ylabel("D(E)")
plt.legend(loc="best")

#print(np.sort(eps_k_mat.flatten()))
```

```python
#plot hofstadter butterfly using eigenvalues
ev_mat = np.empty((N+1,N),dtype=float);
for nflux in np.arange(N+1):
    Hmat = tb.H_periodic_2D(Nx,Ny,t,tp, B = nflux,alpha = alpha)
    evals = np.linalg.eigvalsh(Hmat); 
    ev_mat[nflux-1] = evals
```

```python
plt.figure(figsize=(14,10))
for nflux in np.arange(N+1):
    plt.plot(ev_mat[nflux-1],np.ones(N)*nflux/N,'.',ms='0.5',color='k')
    plt.xlabel("energy eigenvalues")
    plt.ylabel('nflux/N')
    
plt.savefig("hofstadter-butterfly.png",dpi=400,bbox_inches="tight")
```

```python
#TODO: HOW TO IMPLEMENT CORRECT BOUNDARY CONDITION????
def harper_matrix(nu,p,q,s):
    marr = np.arange(q)
    diagarr = 2*np.cos(2*np.pi*p/q*marr + nu);
    return diags([diagarr,np.ones(q),np.ones(q),s,s],[0,-1,1,q-1,-q+1],shape=(q,q),dtype=complex)
```

```python
nu = 8;p = 1;q = N;s = 1;
```

```python
nu_arr = 2*np.pi*np.arange(Ny)/Ny
eps_mat1 = np.empty(N,dtype=float)
eps_mat2 = np.empty(N,dtype=float)
mat_dens1 = harper_matrix(nu,p,q,s).toarray()
eps_mat1, U1 = np.linalg.eigh(mat_dens1); 
mat_dens2 = harper_matrix(nu,p,q,-s).toarray()
eps_mat2, U2 = np.linalg.eigh(mat_dens2); 
assert np.linalg.norm((U1 * eps_mat1) @ (U1.T.conj()) - mat_dens1) < 1e-10
assert np.linalg.norm((U2 * eps_mat2) @ (U2.T.conj()) - mat_dens2) < 1e-10
```

```python
nu_arr = 2*np.pi*np.arange(Ny)/Ny
p_arr = np.arange(0,N+1)
eps_mat1 = np.full((N+1,N),np.nan,dtype=float)
eps_mat2 = np.full((N+1,N),np.nan,dtype=float)
nu = nu_arr[6]
for p in p_arr:
    g = np.gcd(p,q)
    #print(g)
    if g == 1:
        mat_dens = harper_matrix(nu,p,q,1).toarray()
        eps_mat1[p-1] = np.linalg.eigvalsh(mat_dens); 
        mat_dens = harper_matrix(nu,p,q,-1).toarray()
        eps_mat2[p-1] = np.linalg.eigvalsh(mat_dens); 
    elif g >= q//2:
        pp = p//g*4; qq = q//g*4
        #print(pp,qq)
        mat_dens = harper_matrix(nu,pp,qq,1).toarray()
        eps_mat1[p-1,:qq] = np.linalg.eigvalsh(mat_dens); 
        mat_dens = harper_matrix(nu,pp,qq,-1).toarray()
        eps_mat2[p-1,:qq] = np.linalg.eigvalsh(mat_dens); 
    else: 
        pp = p//g; qq = q//g
        #print(pp,qq)
        mat_dens = harper_matrix(nu,pp,qq,1).toarray()
        eps_mat1[p-1,:qq] = np.linalg.eigvalsh(mat_dens); 
        mat_dens = harper_matrix(nu,pp,qq,-1).toarray()
        eps_mat2[p-1,:qq] = np.linalg.eigvalsh(mat_dens); 
```
