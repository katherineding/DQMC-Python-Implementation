# Equal time measurements of size Nx x Ny

```python
plotroot="hallplots/equal-time-from-nersc/"
resultroot="halldata/equal-time-from-nersc/"
meas_list = ["zz",'xx','nn','pair_sw','g00']
```

```python
Nx = "10"; Ny = "10"; nt = "1.0"; tp='0'; N = int(Nx) * int(Ny)
if Nx == '6' and Ny == '6':
    nflux_list = ['0','2','4','6','8','10','12','14','16','32'];
    Us_list =['0','4','6','8']
    betas_list = ['3','4','5','6']
    pathroot = hdpath+"NotOnDell/Data-DQMC/compressibility/6x6_tp0_all/"
elif Nx == '8' and Ny == '8' and tp =='0':
    nflux_list = ['0','1','2','3','4','6','8','12','16','20','24','28','32'];
    Us_list =['0','2','4','6','8']
    betas_list = ['3','4','5','6']
    pathroot = hdpath+"NotOnDell/Data-DQMC/compressibility/"
    #nflux_list = ['0','1'];
    #Us_list =['0']
    #betas_list = ['5']
elif Nx == '8' and Ny == '8' and tp =='-0.25':
    #special: on SSD
    nflux_list = ['0','4','8','12','16','20','24','28','32'];
    Us_list =['0','4','6','8']
    betas_list = ['3','4','5','6']
    pathroot = ssdpath+"Data-DQMC/compressibility/"
elif Nx == "10" and Ny == "10":
    nflux_list = ['0','6','12','18','25','32','37','43','50'];
    Us_list =['0','4','6','8']
    betas_list = ['3','4','5','6']
    pathroot = hdpath+"NotOnDell/Data-DQMC/compressibility/"
nf_arr = np.array(list(map(float,nflux_list)));
B_arr = nf_arr/N; 
nF = len(nflux_list);nT = len(betas_list);nU = len(Us_list)
tt_prefix = f"{Nx}x{Ny} t'={tp}"
```

```python
#try with data generated for compressibility first
print(pathroot)
mu_arr = np.linspace(-6,6,101); nmu = mu_arr.shape[0];
```

```python
print("nflux_list:",nflux_list,"length =",len(nflux_list))
print("betas_list:",betas_list,"length =",nT);
print("Us_list:",Us_list,"length =",nU); 
```

## correlation functions zz, xx, nn, pair_sw, g00

```python
### Method 1 : Dictonary. This reduces file reads, 
###    but only saves progress after all have been completed
###

#initialize dictionaries
data_dict = {}; 
dataerr_dict = {}; 
for meas_name in meas_list:
    data_dict[meas_name] = np.full((nT,nU,nF,nmu,int(Nx),int(Ny)),np.nan,dtype=complex)
    dataerr_dict[meas_name] = np.full((nT,nU,nF,nmu,int(Nx),int(Ny)),np.nan,dtype=float)


for k in range(nF):
    nflux = nflux_list[k]
    for j in range(nU):
        Us = Us_list[j]
        for i in range(nT):
            betas=betas_list[i]
            for m in range(nmu):
                mu = mu_arr[m]
                dpath = pathroot+f"{Nx}x{Ny}_tp{tp}_nflux{nflux}/n{nt}/beta{betas}_U{Us}/mu{mu:.3f}/"
                print(dpath)
                try: 
                    dm, de = da.eqlt_meas_Nx_Ny(dpath,meas_list)
                    for meas_name in meas_list:
                        data_dict[meas_name][i,j,k,m] = dm[meas_name];
                        dataerr_dict[meas_name][i,j,k,m] = de[meas_name];
                except:
                    pass
                
```

```python
for meas_name in meas_list:
    np.save(resultroot+f"{meas_name}/Nx{Nx}_Ny{Ny}_tp{tp}_mu.npy",mu_arr)
    np.save(resultroot+f"{meas_name}/Nx{Nx}_Ny{Ny}_tp{tp}_{meas_name}.npy",data_dict[meas_name])
    np.save(resultroot+f"{meas_name}/Nx{Nx}_Ny{Ny}_tp{tp}_{meas_name}err.npy",dataerr_dict[meas_name])
    np.save(resultroot+f"{meas_name}/Nx{Nx}_Ny{Ny}_tp{tp}_nflux.npy",nflux_list)
    np.save(resultroot+f"{meas_name}/Nx{Nx}_Ny{Ny}_tp{tp}_Us.npy",Us_list)
    np.save(resultroot+f"{meas_name}/Nx{Nx}_Ny{Ny}_tp{tp}_betas.npy",betas_list)
    data_ld = np.load(resultroot+f"{meas_name}/Nx{Nx}_Ny{Ny}_tp{tp}_{meas_name}.npy")
    num_ok = np.count_nonzero(~np.isnan(data_ld))
    print(f"{int(num_ok)}/{data_ld.size} = {int(num_ok)/data_ld.size:.3g} of data is saved in directory {resultroot}{meas_name}/")
    print(f"{meas_name}_ld.shape = {data_ld.shape}")
```
