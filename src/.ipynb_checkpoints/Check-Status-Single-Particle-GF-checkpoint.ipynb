```python
Nx = "6"; Ny = "6"; nflux = "2"; nt="0.95";tp="0" ; N = int(Nx)*int(Ny)
if tp == "0":
    pathroot = f"/home/katherine/Documents/GradWork/Devereaux/Data-DQMC/{Nx}x{Ny}/nflux{nflux}/n{nt}/"
else:
    pathroot = f"/home/katherine/Documents/GradWork/Devereaux/Data-DQMC/{Nx}x{Ny}_tp{tp}/nflux{nflux}/n{nt}/"
tt_prefix = f"{Nx}x{Ny} t'={tp} n={nt}"
```

```python
plt.figure();
plt.title(f'{Nx}x{Ny} nflux={nflux} target n = {nt}')
betas_list, Us_list, mu_dict = da.readmu(pathroot,"best_mu.txt",show=True)
nT = len(betas_list); nU = len(Us_list)
beta_list = np.array(list(map(float,betas_list)))
U_list = np.array(list(map(float,Us_list)))
#reorder beta and U lists
border = np.argsort(beta_list); Uorder = np.argsort(U_list)
#plt.savefig(f'{plotroot}{Nx}x{Ny}_nflux{nflux}_n{nt}_mus.png',dpi=300,bbox_inches="tight")
#plt.close()
```

```python
for i in border:
    betas = betas_list[i]
    for j in Uorder[1:]:
        Us = Us_list[j]
        mu = mu_dict[(betas,Us)]
        path= pathroot + f"beta{betas}_U{Us}_mu{mu:.3f}/"
        try:
            da.info(path,uneqlt=False,show=True,imagtol=1e-2)
        except IndexError as e:
            print(e)
```

```python
for i in border:
    betas = betas_list[i]
    for j in Uorder[1:]:
        Us = Us_list[j]
        mu = mu_dict[(betas,Us)]
        plt.figure();
        plt.title(tt_prefix+fr" B={nflux}/{N} $\beta$={betas} U={Us}")
        try:
            path= pathroot + f"beta{betas}_U{Us}_mu{mu:.3f}/"
            da.plot_gt0_ii(path,nt,spec="real+imag")
        except ValueError:
            path= pathroot + f"beta{betas}_U{Us}_mu{mu:.4f}/"
            da.plot_gt0_ii(path,nt,spec="real+imag")
        #plt.show()
        plt.show()
        #plt.savefig(f'{plotroot}{Nx}x{Ny}_nflux{nflux}_n{nt}_beta{betas}_U{Us}_profiles.png',dpi=300,bbox_inches="tight")
        #plt.close()
```

```python
for i in border:
    betas = betas_list[i]
    plt.figure()
    plt.title(tt_prefix+fr" B={nflux}/{N} $\beta$={betas}")
    for j in Uorder[1:]:
        Us = Us_list[j]
        mu = mu_dict[(betas,Us)]
        try:
            path= pathroot + f"beta{betas}_U{Us}_mu{mu:.3f}/"
            da.plot_gt0_ii(path,nt,spec="Ulabel")
        except ValueError:
            path= pathroot + f"beta{betas}_U{Us}_mu{mu:.4f}/"
            da.plot_gt0_ii(path,nt,spec="Ulabel")
    #plt.savefig(f'{plotroot}{Nx}x{Ny}_nflux{nflux}_n{nt}_beta{betas}_Uvar.png',dpi=300,bbox_inches="tight")
    #plt.close()
```
