### 1D : Can find analytic DOS in TD limit for t = 0

```python
Nx =1000; t = 1; tp = -0.25; N = Nx;
'''Diagonalize H-mat numerically: ED'''
#Hmat = tb.H_periodic_1D(Nx,t,tp)
#evs = eigsh(Hmat,k=Nx-1,return_eigenvectors=False)
'''Known exact eigenvalues, should agree with ED'''
eps_k_arr = tb.eps_k_1D(Nx,t,tp)
'''poisson kernels'''
earr = np.linspace(-3,3,1001); sig=1/50
Dearr = np.zeros(earr.shape,dtype=float)
for i in range(int(Nx)):
    Dearr += tb.poisson(sig,earr,eps_k_arr[i])/Nx
'''1D special case: analytical expression of D(E) is known in TD limit, with t' = 0'''
Dearr_tp0 = 1/np.sqrt(1-np.power(earr/2,2))/(2*np.pi)
```

```python
plt.figure(figsize=(10,8))
plt.hist(eps_k_arr,bins=50,density=True,alpha = 0.9, label="formula eigenvalue hist")
plt.plot(earr,Dearr,label="poisson proxy, formula eigenvals")
#plt.plot(earr,Dearr_tp0,label="exact DOS TD limit t'=0")

plt.title(f"1D Nx={Nx} t'={tp}")
plt.xlabel("energy E/t")
plt.ylabel("D(E)")
plt.legend(loc="best")
```

```python
beta=10
mu_arr = np.linspace(-6,6,101)
#n_arr = np.full(mu_arr.shape,np.nan,dtype=float)
#E_arr = np.full(mu_arr.shape,np.nan,dtype=float)
n_arr = np.sum(tb.f_dist_mat(eps_k_arr,mu_arr,beta),axis=0)/N*2
d = np.sum(tb.f_dist_deriv_mat(eps_k_arr,mu_arr,beta),axis=0)/N*2 #factor of 2 from spin degeneracy
E_arr = np.sum(eps_k_arr[:,np.newaxis] * tb.f_dist_mat(eps_k_arr,mu_arr,beta),axis=0)/N*2
dE_arr = np.sum(eps_k_arr[:,np.newaxis] * tb.f_dist_deriv_mat(eps_k_arr,mu_arr,beta),axis=0)/N*2
```

```python
plt.figure(figsize=(10,8))
plt.plot(mu_arr,n_arr,label = "$n(\mu)$")
plt.plot(mu_arr,d,":",lw=4,label = r"$d\langle n\rangle /d\mu$")
plt.title(fr"1D Nx={Nx} t'={tp} $\beta$={beta}")
plt.xlabel(r"chemical potential $\mu/t$")
plt.legend(loc="best")
plt.grid(True)
```

```python
plt.figure(figsize=(10,8))
plt.plot(mu_arr,E_arr,label = "$E(\mu)$")
plt.plot(mu_arr,dE_arr,":",lw=4,label = r"$d\langle E\rangle /d\mu$")
plt.title(fr"1D Nx={Nx} t'={tp} $\beta$={beta}")
plt.xlabel(r"chemical potential $\mu/t$")
plt.legend(loc="best")
plt.grid(True)
```

## Free electron gas

```python
def freeDOS(E):
    return np.where(E<=0,0,np.sqrt(E))
```

```python
earr = np.linspace(-2,12,4001);
```

```python
beta=10
mu_arr = np.linspace(-2,8,100)
n_arr = np.trapz(tb.f_dist_mat(earr,mu_arr,beta)*freeDOS(earr)[:,None], x=earr,axis=0)
```

```python
plt.figure(figsize=(10,8))
plt.plot(earr,freeDOS(earr),label = "D(E)")
plt.plot(earr,tb.f_dist(earr,0,beta),label = f"f(E,mu = 0,beta={beta})")
plt.plot(mu_arr,n_arr,label = "n(mu)")
plt.plot(mu_arr,2/3*np.power(mu_arr,3/2),':',label="T = 0 n(mu)")
plt.xlim(-2,3)
plt.ylim(-0.1,2)

spl = InterpolatedUnivariateSpline(mu_arr,n_arr,ext='raise',check_finite=True)
d = spl.derivative();
plt.plot(mu_arr,d(mu_arr),label = "dn/dmu")
plt.legend(loc="best")
```
