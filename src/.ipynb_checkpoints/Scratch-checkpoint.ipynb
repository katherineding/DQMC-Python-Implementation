```python
beta=40; mu = 3
nu = lambda x : np.sum(1/(np.exp(beta*(x*(np.arange(50)[:,None]+1/2)-mu))+1),axis=0)
n0 = lambda x : x*np.sum(1/(np.exp(beta*(x*(np.arange(100)[:,None]+1/2)-mu))+1),axis=0)
barr = np.arange(0.03,8,0.001)
plt.figure();plt.plot(barr,nu(barr));plt.ylim(0,6);plt.ylabel(r"filling factor $\nu$");plt.xlabel(r"$B$");plt.grid(True)
plt.figure();plt.plot(barr,n0(barr));plt.ylabel(r"Hall number $n_0$");plt.xlabel(r"$B$");plt.grid(True)
```

## Band structure for 2D tight binding model

```python
def kin_energy(kx,ky):
    return -2*(np.cos(kx) + np.reshape(np.cos(ky),(len(ky),1)))
```

Center of BZ is k = (0,0)

```python
Nx=16
Ny=16
kx_arr = np.linspace(-np.pi,np.pi,Nx+1)
ky_arr = np.linspace(-np.pi,np.pi,Ny+1)

plt.figure()
plt.title(r"$\epsilon_k$ as a function of wave vector")
plt.contourf(kx_arr,ky_arr, kin_energy(kx_arr,ky_arr));plt.colorbar();
plt.xlim(-np.pi,np.pi);plt.ylim(-np.pi,np.pi);
plt.xlabel(r"$k_x$");plt.ylabel(r"$k_y$");
plt.xticks(np.arange(-np.pi, 1.5*np.pi, step=0.5*np.pi),\
           (r'$-\pi$', r'$-\pi/2$',r'$0$',r'$\pi/2$', r'$\pi$'));
plt.yticks(np.arange(-np.pi, 1.5*np.pi, step=0.5*np.pi),\
           (r'$-\pi$', r'$-\pi/2$',r'$0$',r'$\pi/2$', r'$\pi$'));
plt.axis('scaled');
plt.grid(True)
```

Hole Hamiltonian inverts bands, forcing us to choose new center for BZ as $\vec{k} = (\pi,\pi)$

```python
Nx=16
Ny=16
kx_arr = np.linspace(0,2*np.pi,Nx+1)
ky_arr = np.linspace(0,2*np.pi,Ny+1)

plt.figure()
plt.title(r"$\epsilon_k$ as a function of wave vector")
plt.contourf(kx_arr,ky_arr, -kin_energy(kx_arr,ky_arr));plt.colorbar();
plt.xlim(0,2*np.pi);plt.ylim(0,2*np.pi);
plt.xlabel(r"$k_x$");plt.ylabel(r"$k_y$");
plt.xticks(np.arange(0, 2.5*np.pi, step=0.5*np.pi),\
           (r'$0$', r'$\pi/2$',r'$\pi$',r'$3\pi/2$', r'$2\pi$'));
plt.yticks(np.arange(0, 2.5*np.pi, step=0.5*np.pi),\
           (r'$0$', r'$\pi/2$',r'$\pi$',r'$3\pi/2$', r'$2\pi$'));
plt.axis('scaled');
plt.grid(True)
```
